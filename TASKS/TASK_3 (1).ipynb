**Week 3: Univariate and Bivariate Visual Analysis**

**1.Distribution of Daily Hours (Avg_Daily_Screen_Time_hr**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"/content/Indian_Kids_Screen_Time_cleaned (1).csv")

# Create a 2x2 grid of subplots
plt.figure(figsize=(12, 10))

# Histogram
plt.subplot(2, 2, 1)
plt.hist(df['Avg_Daily_Screen_Time_hr'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram - Daily Hours")
plt.xlabel("Hours")
plt.ylabel("Frequency")

#  Boxplot
plt.subplot(2, 2, 2)
sns.boxplot(y=df['Avg_Daily_Screen_Time_hr'], color="lightgreen")
plt.title("Boxplot - Daily Hours")

#  Bar Chart (rounded)
plt.subplot(2, 2, 3)
df['Avg_Daily_Screen_Time_hr'].round().value_counts().sort_index().plot(kind='bar', color="orange")
plt.title("Bar Chart - Daily Hours (rounded)")
plt.xlabel("Hours")
plt.ylabel("Count")

#  Pie Chart (rounded)
plt.subplot(2, 2, 4)
df['Avg_Daily_Screen_Time_hr'].round().value_counts().plot(kind='pie', autopct='%1.1f%%', cmap="Set3")
plt.title("Pie Chart - Daily Hours (rounded)")
plt.ylabel("")

# Adjust layout and show all plots together
plt.tight_layout()
plt.show()
```

**2. Distribution of Age Band (Age_Band)**

```python
# Create a 1x3 grid of subplots
plt.figure(figsize=(15, 5))

#  Bar Chart
plt.subplot(1, 3, 1)
df['Age_Band'].value_counts().plot(kind='bar', color="skyblue")
plt.title("Bar Chart - Age Band")
plt.xlabel("Age Band")
plt.ylabel("Count")

#  Pie Chart
plt.subplot(1, 3, 2)
df['Age_Band'].value_counts().plot(kind='pie', autopct='%1.1f%%', cmap="Set2")
plt.title("Pie Chart - Age Band")
plt.ylabel("")

# Heatmap (counts)
plt.subplot(1, 3, 3)
sns.heatmap(pd.crosstab(index=df['Age_Band'], columns="count"), annot=True, fmt="d", cmap="Blues")
plt.title("Heatmap - Age Band")

# Adjust layout and display all three plots together
plt.tight_layout()
plt.show()
```

**3. Distribution of Device Usage (Primary_Device or Device_Type)**

```python
# Create a 1x3 grid of subplots
plt.figure(figsize=(15, 5))

#  Bar Chart
plt.subplot(1, 3, 1)
df['Primary_Device'].value_counts().plot(kind='bar', color="lightcoral")
plt.title("Bar Chart - Primary Device")
plt.xlabel("Device")
plt.ylabel("Count")

#  Pie Chart
plt.subplot(1, 3, 2)
df['Primary_Device'].value_counts().plot(kind='pie', autopct='%1.1f%%', cmap="Set1")
plt.title("Pie Chart - Primary Device")
plt.ylabel("")

#  Heatmap (counts)
plt.subplot(1, 3, 3)
sns.heatmap(pd.crosstab(index=df['Primary_Device'], columns="count"), annot=True, fmt="d", cmap="Greens")
plt.title("Heatmap - Primary Device")

# Adjust layout and show all three together
plt.tight_layout()
plt.show()
```

**5. Compare Screen Time by Age Band**

```python
# Create a 2x2 grid of subplots
plt.figure(figsize=(12, 10))

#  Boxplot
plt.subplot(2, 2, 1)
sns.boxplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title("Boxplot - Screen Time by Age Band")

#  Bar Chart (mean)
plt.subplot(2, 2, 2)
sns.barplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean')
plt.title("Bar Chart (Mean) - Screen Time by Age Band")

#  Line Plot (mean)
plt.subplot(2, 2, 3)
sns.lineplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean', marker='o')
plt.title("Line Plot - Screen Time by Age Band")

#  Heatmap (mean)
plt.subplot(2, 2, 4)
sns.heatmap(
    pd.pivot_table(df, values='Avg_Daily_Screen_Time_hr', index='Age_Band', aggfunc='mean').T,
    annot=True, cmap="YlGnBu"
)
plt.title("Heatmap - Avg Screen Time by Age Band")

# Adjust layout and show all 4 together
plt.tight_layout()
plt.show()
```

**6. Compare Screen Time by Location Type (Urban_or_Rural)**

```python

# Create a 2x2 grid of subplots
plt.figure(figsize=(12, 10))

#  Boxplot
plt.subplot(2, 2, 1)
sns.boxplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title("Boxplot - Screen Time by Location Type")

# Bar Chart (mean)
plt.subplot(2, 2, 2)
sns.barplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean')
plt.title("Bar Chart (Mean) - Screen Time by Location Type")

#  Line Plot (mean)
plt.subplot(2, 2, 3)
sns.lineplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean', marker='o')
plt.title("Line Plot - Screen Time by Location Type")

#  Heatmap (mean)
plt.subplot(2, 2, 4)
sns.heatmap(
    pd.pivot_table(df, values='Avg_Daily_Screen_Time_hr', index='Urban_or_Rural', aggfunc='mean').T,
    annot=True, cmap="OrRd"
)
plt.title("Heatmap - Avg Screen Time by Location Type")

# Adjust layout and show all together
plt.tight_layout()
plt.show()
```
