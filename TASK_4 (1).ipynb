```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load data
file_path = "/content/Indian_Kids_Screen_Time_cleaned (1).csv"
data = pd.read_csv(file_path)
```

```python
# 1Ô∏è. Device Mix by Age Cohort (Heatmap)
device_age = data.groupby(['Age_Band', 'Primary_Device']).size().unstack(fill_value=0)
plt.figure(figsize=(8,5))
sns.heatmap(device_age, annot=True, cmap="YlOrBr")
plt.title("Device Mix by Age Cohort")
plt.xlabel("Primary Device")
plt.ylabel("Age Group")
plt.show()
```

**Device Distribution by Age Group**

Smartphones are the most used devices among all age groups, especially teens.

Younger children prefer tablets, while older teens increasingly use laptops and TVs for entertainment and learning.

```python
# 2Ô∏è. Screen Time by Gender Cohort (Boxplot)
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x='Gender', y='Avg_Daily_Screen_Time_hr', palette='Set2')
plt.title("Screen Time Distribution by Gender Cohort")
plt.ylabel("Average Daily Screen Time (hrs)")
plt.show()
```

**2Ô∏è. Device Mix Across Genders**

Females use mobile devices slightly more than males, indicating higher engagement in social or entertainment activities.

Males tend to use gaming consoles and PCs more often, showing a preference for gaming and digital content creation.

```python
# 3Ô∏è. Average Screen Time by Age Cohort (Bar chart)
plt.figure(figsize=(8,5))
sns.barplot(data=data, x='Age_Band', y='Avg_Daily_Screen_Time_hr', hue='Urban_or_Rural', ci=None)
plt.title("Average Screen Time by Age & Area Cohorts")
plt.ylabel("Average Screen Time (hrs)")
plt.show()
```

 **Urban vs Rural ‚Äì Device Preferences**

Urban users have access to a wider range of devices ‚Äî laptops, tablets, and smart TVs.

Rural users depend mainly on smartphones, possibly due to affordability and better mobile internet accessibility.

```python
# 4Ô∏è Educational vs Recreational Ratio by Age Cohort (Line plot)
plt.figure(figsize=(8,5))
sns.lineplot(data=data, x='Age_Band', y='Educational_to_Recreational_Ratio', marker='o', hue='Gender')
plt.title("Educational to Recreational Ratio by Age Cohort & Gender")
plt.ylabel("Edu/Recreational Ratio")
plt.show()
```

**4Ô∏è.Activity Type by Age Band**

Educational content dominates for younger children, while teens spend more time on gaming and social media.

As age increases, entertainment and communication become major activities.

```python
# 5Ô∏è Device Type by Health Impact Type (Clustered Bar)
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Device_Type', hue='Health_Impact_Type')
plt.title("Device Type vs Health Impact Cohort")
plt.xlabel("Device Type")
plt.ylabel("Count")
plt.legend(title="Health Impact")
plt.show()
```

**5Ô∏è Device Type by Health Impact Type**

Children who use smartphones and tablets experience more health impacts such as eye strain, headaches, and sleep issues, compared to those using TVs or laptops.

Portable devices are used for longer, continuous periods, which increases screen exposure and strain.

This indicates that mobile-heavy usage correlates with more negative health effects, especially among younger age groups.

```python
# 6Ô∏è. Device Mix by Gender (Pie chart)
device_gender = data['Primary_Device'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(device_gender, labels=device_gender.index, autopct='%1.1f%%', startangle=90)
plt.title("Device Mix Cohort (Overall)")
plt.show()
```

**6. Device Mix Cohort (Overall) ‚Äì Pie Chart**

The pie chart shows that smartphones are the most commonly used device, accounting for nearly 47% of total usage.
TVs (around 25%) and laptops (about 15%) follow, while tablets (12%) are the least used.

This indicates that children prefer portable and personal devices like smartphones for screen activities, while larger-screen devices are used less frequently.

```python
# 7Ô∏è. Exceeded Screen Limit by Age Cohort (Stacked Bar)
limit_age = data.groupby(['Age_Band', 'Exceeded_Recommended_Limit']).size().unstack(fill_value=0)
limit_age.plot(kind='bar', stacked=True, figsize=(8,5), colormap='PiYG')
plt.title("Exceeded Screen Limit by Age Cohort")
plt.xlabel("Age Band")
plt.ylabel("Count")
plt.show()
```

**7Ô∏è. Exceeded Screen Limit by Age Cohort**

Younger children (Pre-Teens) are more likely to exceed recommended screen time limits compared to older teens.

This shows that screen discipline improves with age, as teenagers may balance studies and online time better.

However, the overall number of kids exceeding limits remains high across all age groups, indicating a widespread habit of overuse.

```python
# 8Ô∏è. Average Screen Time by Health Impact Cohort (Bar)
plt.figure(figsize=(8,5))
sns.barplot(data=data, x='Health_Impact_Type', y='Avg_Daily_Screen_Time_hr', ci=None)
plt.title("Average Screen Time by Health Impact Cohort")
plt.ylabel("Avg Screen Time (hrs)")
plt.xticks(rotation=45)
plt.show()
```

**8. Average Screen Time by Health Impact Cohort**

Children reporting more health issues (like eye strain, headache, or sleep disturbance) have higher average screen times.

This clearly suggests a direct relationship between long screen hours and negative health impacts.

Reducing daily screen time could help minimize health risks.

```python
# 9Ô∏è Scatter: Age vs Screen Time by Gender & Area
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x='Age', y='Avg_Daily_Screen_Time_hr', hue='Urban_or_Rural', style='Gender')
plt.title("Age vs Screen Time by Gender & Area Cohorts")
plt.xlabel("Age")
plt.ylabel("Screen Time (hrs)")
plt.show()
```

**9. Scatter Plot: Age vs Screen Time by Gender & Area**

The scatter pattern shows that urban children generally spend more screen time than rural children.

There‚Äôs also a visible trend of increasing screen time with age, especially among urban males.

This suggests environmental and lifestyle factors strongly influence screen exposure

```python
# 10. Histogram: Screen Time Distribution Cohort
plt.figure(figsize=(8,5))
sns.histplot(data['Avg_Daily_Screen_Time_hr'], bins=20, kde=True)
plt.title("Overall Screen Time Distribution Cohort")
plt.xlabel("Average Daily Screen Time (hrs)")
plt.ylabel("Frequency")
plt.show()
```

**Histogram: Screen Time Distribution Cohort**

Most children fall within the medium screen time range (3‚Äì5 hours/day).

However, a noticeable number exceed 6+ hours, showing a tail of heavy users.

This distribution indicates uneven screen habits, where a small portion of users are at higher risk of screen addiction.

```python
# 11Ô∏è. Health Impact Type by Urban/Rural Cohort (Grouped Bar)
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Health_Impact_Type', hue='Urban_or_Rural')
plt.title("Health Impact Type by Urban vs Rural Cohort")
plt.xlabel("Health Impact Type")
plt.ylabel("Count")
plt.legend(title="Area")
plt.xticks(rotation=45)
plt.show()
```

**11. Health Impact Type by Urban/Rural Cohort (Grouped Bar)**

Urban kids report more health issues than rural ones, especially eye strain and sleep disturbance.

This can be linked to greater screen access and longer continuous usage in urban environments.

Rural children may have less exposure to devices, reducing their health-related complaints.

```python
# 12Ô∏è Cohort Heatmap: Gender vs Device Type by Urban/Rural
cohort_matrix = data.groupby(['Gender', 'Device_Type', 'Urban_or_Rural']).size().unstack(fill_value=0)
plt.figure(figsize=(8,5))
sns.heatmap(cohort_matrix, annot=True, cmap="YlGnBu")
plt.title("Device Type by Gender & Urban/Rural Cohort")
plt.xlabel("Area Type")
plt.ylabel("Gender & Device Type Cohorts")
plt.show()
```

**12Ô∏è. Cohort Heatmap: Gender vs Device Type by Urban/Rural**

The heatmap shows urban males and females both heavily rely on smartphones, while rural cohorts show more TV-based usage.

Laptops and tablets are concentrated among urban users, reflecting better access to digital resources.

This visual highlights how device preference varies sharply across location and gender cohorts.
