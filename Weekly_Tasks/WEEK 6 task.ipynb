```python
# ---------------------------------------------------------
# Step 1: Import libraries and dataset
# ---------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
sns.set(style="whitegrid", context="notebook")

# Load dataset
data = pd.read_csv("Indian_Kids_Screen_Time_Updated.csv")
data.head()
```

```python
# SEGMENT-WISE SUMMARY TABLES AND VISUALS
```

```python
# ---------------------------------------------------------
# Segment 1: Summary by Age Band
# ---------------------------------------------------------
age_summary = data.groupby('Age_Band')['Avg_Daily_Screen_Time_hr'].agg(['mean', 'min', 'max', 'count']).reset_index()
print(age_summary)

plt.figure(figsize=(8,5))
sns.barplot(x='Age_Band', y='mean', data=age_summary, palette='pastel')
plt.title('Average Screen Time by Age Band')
plt.xlabel('Age Band')
plt.ylabel('Average Screen Time (hrs)')
plt.show()
```

```python
OBSERVATION:
1. Teenagers show the highest mean screen time across all groups.
2. Pre-teens follow, with a moderate average, while children have the lowest.
3. This clearly indicates that age maturity increases digital involvement.
4. The teenage cohort is the top contributor to high screen time behavior.
```

```python
# ---------------------------------------------------------
# Segment 2: Summary by Gender
# ---------------------------------------------------------
gender_summary = data.groupby('Gender')['Avg_Daily_Screen_Time_hr'].agg(['mean', 'min', 'max', 'count']).reset_index()
print(gender_summary)

plt.figure(figsize=(6,5))
sns.barplot(x='Gender', y='mean', data=gender_summary, palette='Set2')
plt.title('Average Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Screen Time (hrs)')
plt.show()
```

```python
OBSERVATION:
1. Male participants record slightly higher screen usage than females.
2. Both genders, however, spend considerable time online, averaging over 3 hours.
3. Male dominance may be tied to gaming or entertainment-driven activities.
4. Gender differences exist but are less significant compared to age and device type.
```

```python
# ---------------------------------------------------------
# Segment 3: Summary by Urban vs Rural
# ---------------------------------------------------------
region_summary = data.groupby('Urban_or_Rural')['Avg_Daily_Screen_Time_hr'].agg(['mean', 'min', 'max', 'count']).reset_index()
print(region_summary)

plt.figure(figsize=(6,5))
sns.barplot(x='Urban_or_Rural', y='mean', data=region_summary, palette='coolwarm')
plt.title('Average Screen Time by Location Type')
plt.xlabel('Region Type')
plt.ylabel('Average Screen Time (hrs)')
plt.show()
```

```python
OBSERVATION:
1. Urban children have noticeably higher average screen time compared to rural ones.
2. Urban regions provide more device access and stronger digital exposure.
3. Rural children spend less time, likely due to limited access or outdoor play.
4. This geographic difference reveals environment as a key external factor.
```

```python
# ---------------------------------------------------------
# Segment 4: Summary by Device Type
# ---------------------------------------------------------
device_summary = data.groupby('Primary_Device')['Avg_Daily_Screen_Time_hr'].agg(['mean', 'min', 'max', 'count']).reset_index()
print(device_summary)

plt.figure(figsize=(8,5))
sns.barplot(x='Primary_Device', y='mean', data=device_summary, palette='viridis')
plt.title('Average Screen Time by Device Type')
plt.xlabel('Primary Device')
plt.ylabel('Average Screen Time (hrs)')
plt.show()
```

```python
OBSERVATION:
1. Smartphone users record the highest average screen hours, often exceeding 4 hours.
2. Tablet users follow, while desktop or console users show lower averages.
3. Portable devices drive higher engagement due to convenience and multitasking.
4. Hence, device type is one of the strongest behavioral screen time drivers.
```

```python
# ---------------------------------------------------------
# Segment 5: Health Category vs Average Screen Time
# ---------------------------------------------------------
health_summary = data.groupby('Health_Category')['Avg_Daily_Screen_Time_hr'].agg(['mean', 'min', 'max', 'count']).reset_index()
print(health_summary)

plt.figure(figsize=(8,5))
sns.barplot(x='Health_Category', y='mean', data=health_summary, palette='mako')
plt.title('Average Screen Time by Health Category')
plt.xlabel('Health Category')
plt.ylabel('Average Screen Time (hrs)')
plt.show()
```

```python
OBSERVATION:
1. Physical health-related impacts correspond to higher average screen time.
2. No Impact users maintain a healthy, controlled screen duration.
3. High screen hours directly increase the risk of eye strain or obesity.
4. This reinforces screen time as a predictor of physical health effects.
```

```python
# ---------------------------------------------------------
# Segment 6: Educational vs Recreational Ratio and Screen Time
# ---------------------------------------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x='Educational_to_Recreational_Ratio', y='Avg_Daily_Screen_Time_hr',
                hue='Age_Band', data=data, palette='Set1')
plt.title('Educational to Recreational Ratio vs Screen Time')
plt.xlabel('Educational to Recreational Ratio')
plt.ylabel('Average Screen Time (hrs)')
plt.show()
```

```python
OBSERVATION:
1. A lower educational-to-recreational ratio indicates more recreational usage and higher screen time.
2. Higher ratios (more educational) correspond to balanced, moderate usage.
3. Teenagers dominate the low-ratio, high-usage zone.
4. Thus, the purpose of device usage critically defines digital wellness.
```

```python
# ---------------------------------------------------------
# Segment 7: Exceeded Limit Distribution
# ---------------------------------------------------------
plt.figure(figsize=(8,5))
sns.countplot(x='Age_Band', hue='Exceeded_Recommended_Limit', data=data, palette='Set3')
plt.title('Exceeded Recommended Screen Time by Age Band')
plt.xlabel('Age Band')
plt.ylabel('Count of Users')
plt.legend(title='Exceeded Limit')
plt.show()
```

```python
OBSERVATION:
1. Most teenagers have exceeded their recommended daily screen limit.
2. Pre-teens also show increasing numbers nearing unhealthy thresholds.
3. Children, however, mostly remain within safe limits.
4. Hence, exceeding limits is predominantly a teenage concern.
```

```python
# ---------------------------------------------------------
# Segment Summary Table
# ---------------------------------------------------------
segment_summary = data.groupby(['Age_Band', 'Gender', 'Urban_or_Rural'])['Avg_Daily_Screen_Time_hr'].mean().reset_index()
segment_summary = segment_summary.sort_values(by='Avg_Daily_Screen_Time_hr', ascending=False)
print(segment_summary.head(10))
```

```python
# OBSERVATIONS:
1.Teenagers (1319 years) consistently showed higher screen usage compared to Pre-teens (612 years), with most exceeding the recommended screen time limit.

2.Male participants had slightly higher average daily screen time than females across both age bands.

3.Urban children reported more screen time compared to rural children, likely due to greater access to digital devices and internet.

4.Smartphones dominated as the primary device and were highly correlated with elevated screen time hours, especially among urban teenagers.

5.Physical health concerns such as eye strain and obesity risk were more frequent in cohorts with higher usage.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Define High and Low Screen Time Groups ---
avg_screen_time = data['Avg_Daily_Screen_Time_hr'].mean()

high_users = data[data['Avg_Daily_Screen_Time_hr'] > avg_screen_time]
low_users = data[data['Avg_Daily_Screen_Time_hr'] <= avg_screen_time]

print("Average Screen Time across all users:", round(avg_screen_time, 2), "hrs\n")
print("High Screen Time Group:", high_users.shape[0], "users")
print("Low Screen Time Group:", low_users.shape[0], "users")

# --- 2. Summary Statistics ---
print("\nHigh Screen Time Users Summary:")
print(high_users[['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']].describe())

print("\nLow Screen Time Users Summary:")
print(low_users[['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']].describe())

# --- 3. Visualization 1: Average Screen Time by Age Band ---
plt.figure(figsize=(8,5))
sns.barplot(data=data, x='Age_Band', y='Avg_Daily_Screen_Time_hr', hue='Gender', palette='coolwarm')
plt.title('Average Screen Time by Age Band and Gender')
plt.xlabel('Age Band')
plt.ylabel('Average Daily Screen Time (hrs)')
plt.show()

# --- 4. Visualization 2: Screen Time by Device Category ---
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x='Device_Category', y='Avg_Daily_Screen_Time_hr', palette='viridis')
plt.title('Screen Time Distribution Across Device Categories')
plt.xlabel('Device Category')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.show()

# --- 5. Visualization 3: Screen Time vs Educational/Recreational Ratio ---
plt.figure(figsize=(7,5))
sns.scatterplot(data=data, x='Educational_to_Recreational_Ratio', y='Avg_Daily_Screen_Time_hr', hue='Usage_Level', palette='Spectral')
plt.title('Screen Time vs Educational-to-Recreational Ratio')
plt.xlabel('Educational to Recreational Ratio')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.show()

# --- 6. Visualization 4: Screen Time by Health Category ---
plt.figure(figsize=(7,5))
sns.barplot(data=data, x='Health_Category', y='Avg_Daily_Screen_Time_hr', palette='Set2')
plt.title('Average Screen Time by Health Impact Category')
plt.xlabel('Health Category')
plt.ylabel('Average Daily Screen Time (hrs)')
plt.show()

# --- 7. Visualization 5: Urban vs Rural ---
plt.figure(figsize=(6,4))
sns.boxplot(data=data, x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', palette='Accent')
plt.title('Screen Time Comparison: Urban vs Rural Users')
plt.xlabel('Region Type')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.show()
```

```python
OBSERVATIONS:

1.Age and Gender Influence:
Teenagers and young adults (especially 1318 age band) show the highest average screen hours.
Males slightly exceed females in every age group, indicating more gaming or entertainment use.

2.Device Category:
Tablets and smartphones record higher usage durations than laptops or TVs, suggesting mobility and personal device use lead to longer engagement.

3.Educational-to-Recreational Ratio:
Users with a lower ratio (more recreational time) have higher screen usage.
Educationally driven users maintain lower screen exposure  a sign of better control.

4.Health Category:
Higher screen hours are reported in Moderate and Severe health impact groups, showing that excessive use correlates with strain and fatigue.

5.Urban vs Rural Users:
Urban children show significantly higher screen exposure, likely due to better internet access and higher device penetration.
```
