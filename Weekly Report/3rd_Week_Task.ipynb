```python
# ===================================================
# Age Band
# ===================================================

# --- Bar Chart ---
plt.figure(figsize=(7,5))
sns.countplot(x='Age_Band', data=data, palette="pastel")
plt.title("Bar Chart - Age Band")
plt.xlabel("Age Band")
plt.ylabel("Count")
plt.show()

# --- Countplot (same as histogram for categorical) ---
plt.figure(figsize=(7,5))
data['Age_Band'].value_counts().plot(kind="bar", color="coral")
plt.title("Countplot - Age Band")
plt.xlabel("Age Band")
plt.ylabel("Frequency")
plt.show()

# --- Box Plot (Age distribution grouped by Age_Band) ---
plt.figure(figsize=(7,5))
sns.boxplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=data, palette="coolwarm")
plt.title("Boxplot - Screen Time vs Age Band")
plt.show()

# --- Pie Chart ---
plt.figure(figsize=(7,7))
data['Age_Band'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=plt.cm.Set2.colors)
plt.title("Pie Chart - Age Band Distribution")
plt.ylabel("")
plt.show()
```

```python
# ===================================================
# Primary Device
# ===================================================

# --- Bar Chart ---
plt.figure(figsize=(7,5))
sns.countplot(x='Primary_Device', data=data, palette="muted")
plt.title("Bar Chart - Primary Device")
plt.xlabel("Device")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# --- Countplot (alternative bar view) ---
plt.figure(figsize=(7,5))
data['Primary_Device'].value_counts().plot(kind="bar", color="lightgreen")
plt.title("Countplot - Primary Device")
plt.xlabel("Device")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()

# --- Pie Chart ---
plt.figure(figsize=(7,7))
data['Primary_Device'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Pie Chart - Primary Device Distribution")
plt.ylabel("")
plt.show()
```

```python
# ===================================================
#  Screen Time vs Age Band
# ===================================================

plt.figure(figsize=(6,4))
sns.boxplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=data, palette="Set3")
plt.title("Boxplot - Screen Time vs Age Band")
plt.show()

plt.figure(figsize=(6,4))
sns.violinplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=data, palette="husl")
plt.title("Violin Plot - Screen Time vs Age Band")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=data, palette="Blues_d", estimator='mean')
plt.title("Bar Plot (Mean Screen Time) - Age Band")
plt.show()
```
