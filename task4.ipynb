```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
data = pd.read_csv(r"/content/Indian_Kids_Screen_Time_cleaned (1).csv")

# Display first few rows to understand the data
data.head()
```

**Compare device mix and activity categories across demographics**

```python
# 1. Device mix by Gender
plt.figure(figsize=(8,5))
sns.countplot(data=data, x="Primary_Device", hue="Gender")
plt.title("Device Mix by Gender")
plt.xticks(rotation=45)
plt.show()

```

**Understanding**:

Smartphones are the most used device among both boys and girls.

However, girls show slightly higher usage of laptops

```python
# 2 Device mix by Urban/Rural
plt.figure(figsize=(8,5))
sns.countplot(data=data, x="Primary_Device", hue="Urban_or_Rural")
plt.title("Device Mix by Urban/Rural Area")
plt.xticks(rotation=45)
plt.show()
```

**understandings**

Urban children use a wider range of devices  smartphones, laptops, and tablets are more common.

Rural children depend more on TVs and shared family smartphones.

This indicates better device availability and internet access in urban areas.

```python
# 3 Average Screen Time by Age Band
plt.figure(figsize=(8,5))
sns.barplot(data=data, x="Age_Band", y="Avg_Daily_Screen_Time_hr", hue="Gender", ci=None)
plt.title("Average Screen Time by Age Band and Gender")
plt.show()
```

**understanings**

Screen time increases with age  teenagers and late teens spend more hours daily than pre-teens.

Female teens tend to spend slightly more screen time than males in the same age group.

The difference between genders becomes more visible in older age groups.

```python
# 4 Educational vs Recreational Ratio by Age Band
plt.figure(figsize=(8,5))
sns.barplot(data=data, x="Age_Band", y="Educational_to_Recreational_Ratio", hue="Urban_or_Rural", ci=None)
plt.title("Educational vs Recreational Ratio across Age Bands")
plt.show()
```

**understandings**
Pre-teens have a higher educational-to-recreational ratio, meaning more learning-related screen use.

Late teens show a drop in this ratio, meaning screen time is more recreational.

Urban students have a slightly higher educational ratio, suggesting better use of digital learning tools.

```python
# 5. Device Type distribution across Age Bands
# 8 <= age <= 12: PREE TEENS
#  13 <= age <= 16: TEENAGERS
# 17 <= age <= 18: LATE TEENS
plt.figure(figsize=(8,5))
sns.countplot(data=data, x="Primary_Device", hue="Age_Band")
plt.title("Device Type Distribution across Age Bands")
plt.show()
```

**understanings**

Portable devices (smartphones, laptops) are dominant for all age groups, especially for teenagers.

Wall-mounted devices (TVs) are more common among younger kids (pre-teens).

This reflects the trend of mobile-first entertainment and learning among older students

```python
# 6 Health Impact Type by Gender
plt.figure(figsize=(8,5))
sns.countplot(data=data, x="Health_Impact_Type", hue="Gender")
plt.title("Health Impact Types by Gender")
plt.show()
```

**understandings**

Physical impacts (eye strain, poor sleep) are most common for both genders.

Mental impacts (anxiety, low focus) appear slightly more among female students.

This suggests that increased screen exposure affects both health and emotional well-being

```python
# 7 Avg Screen Time by Device Type
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x="Primary_Device", y="Avg_Daily_Screen_Time_hr", hue="Urban_or_Rural")
plt.title("Screen Time by Device Type and Location")
plt.show()
```

**understanings**

Urban users spend more average hours on portable devices compared to rural users.

TV usage shows less variation in time across locations.

The wider range of screen time in urban areas shows different lifestyle patterns and device dependency.

```python
#  (Educational vs Recreational by Device)
# Calculate average Educational % by Device Type
device_usage = data.groupby("Primary_Device")["Edu_Recreational_Percent"].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=device_usage, x="Primary_Device", y="Edu_Recreational_Percent", palette="coolwarm")
plt.title("Average Educational/Recreational Use by Device Type")
plt.ylabel("Average Educational Use (%)")
plt.xlabel("Device Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**understandings**                                 
Laptops and tablets have higher educational usage percentages, meaning they are used more for studying and online learning.

Smartphones and TVs have lower educational use, meaning they are mainly used for recreational purposes like gaming or watching videos.

This clearly shows the purpose-based difference between devices used by kids.
