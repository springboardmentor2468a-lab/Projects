```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Indian_Kids_Screen_Time_Updated.csv")
df.head(7)
```

```python
# Countplot - Age_group distribution
plt.figure(figsize=(7,5))
sns.countplot(x="Age", data=df, color="lightgreen")  
plt.title("Distribution of Age Groups")
plt.show()
```

```python
Insights:1.Uniform Distribution: The most prominent insight is that the distribution of ages is relatively uniform or flat across the range shown (from age 8 to age 18).
```

```python
# Pie chart-Distribution of Device Type
plt.figure(figsize=(6,6))
df['Primary_Device'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    colors=["#66c2a5", "#fc8d62"] 
)

plt.ylabel("")
plt.title("Distribution of Device Type")
plt.show()
```

```python
Insights:Dominance of Smartphones: Smartphones hold the largest share by a significant margin, accounting for 47.0% of the total. This suggests that the Smartphone category is the most popular, most frequently used, or holds nearly half of the market share represented i
```

```python
# countplot-Average Screen Time by Age Group
plt.figure(figsize=(7,5))
sns.barplot(
    x="Age", 
    y="Avg_Daily_Screen_Time_hr", 
    data=df, 
    palette="Set2", 
    hue ="Age",
    errorbar=None
)

plt.title("Average Screen Time by Age Group")
plt.show()
```

```python
Insights:​General Trend: The average daily screen time is generally high and consistent across most age groups, with almost all ages averaging over 3.5 hours per day
```

```python
#  Heatmap - Age_group × Primary_Device
ct = pd.crosstab(df["Age"], df["Primary_Device"])

plt.figure(figsize=(7,5))
sns.heatmap(
    ct,
    annot=True,         
    fmt="d",            
    cmap="YlOrBr",     
    linewidths=1,       
    linecolor="white",  
    cbar=True           
)

plt.title("Device Mix by Age Cohort")
plt.xlabel("Primary Device")
plt.ylabel("Age Group")
plt.show()
```

```python
Insights:Dominance of Smartphones: The Smartphone column consistently shows the highest count values (darkest colors in the middle column) across almost all age groups. This confirms that the Smartphone is the overall most common primary device in this dataset.
```

```python
# Barplot - Health_Status vs Screen Time
plt.figure(figsize=(7,5))
sns.barplot(
    x="Health_Category",
    y="Avg_Daily_Screen_Time_hr",
    data=df,
    palette="Set2",
    errorbar=None,
    hue="Health_Category",          
)

plt.title("Avg Screen Time by Health Status")
plt.show()
```

```python
Insights:​Highest Average Screen Time: The groups reporting a "Physical Impact" and "Mental Impact" from screen time have the highest average daily screen time, both close to 5 hours.
```

```python
# Boxplot - Health_Status vs Edu/Recreational Ratio
plt.figure(figsize=(7,5))
sns.boxplot(
    x="Health_Category", 
    y="Educational_to_Recreational_Ratio", 
    hue="Health_Category",        
    data=df, 
    palette="Set2",               
)
plt.title("Educational vs Recreational Ratio by Health Status")
plt.show()
```

```python
​Highest Average Screen Time: The groups reporting a "Physical Impact" and "Mental Impact" from screen time have the highest average daily screen time, both close to 5 hours.Insights:
```
