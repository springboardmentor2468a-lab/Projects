```python
import pandas as pd

data = pd.read_csv("Indian_Kids_Screen_Time_Updated.csv")

data = data.rename(columns={'Age_Group': 'Age_Band'})
print(data.head())

data.head(4)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ===================================================
# Screen Time
# ===================================================

# --- Bar Chart ---
plt.figure(figsize=(6,4))
data['Avg_Daily_Screen_Time_hr'].value_counts().sort_index().plot(kind='bar', color="skyblue")
plt.title("Bar Chart - Avg Daily Screen Time (hrs)")
plt.xlabel("Screen Time (hrs)")
plt.ylabel("Count")
plt.show()

# --- Histogram ---
plt.figure(figsize=(6,4))
sns.histplot(data['Avg_Daily_Screen_Time_hr'], bins=15, kde=True, color="lightgreen")
plt.title("Histogram - Avg Daily Screen Time (hrs)")
plt.xlabel("Screen Time (hrs)")
plt.show()

# --- Box Plot ---
plt.figure(figsize=(6,4))
sns.boxplot(x=data['Avg_Daily_Screen_Time_hr'], color="orange")
plt.title("Boxplot - Avg Daily Screen Time (hrs)")
plt.show()

# --- Line Plot ---
plt.figure(figsize=(6,4))
sns.lineplot(data=data['Avg_Daily_Screen_Time_hr'], marker="o", color="purple")
plt.title("Line Plot - Avg Daily Screen Time (hrs)")
plt.ylabel("Screen Time (hrs)")
plt.show()

```

```python
# ===================================================
# Age Group
# ===================================================

# --- Bar Chart ---
plt.figure(figsize=(6,4))
data['Age_Band'].value_counts().plot(kind='bar', color="skyblue")
plt.title("Bar Chart - Age Band")
plt.xlabel("Age_Band")
plt.ylabel("Count")
plt.show()

# --- Histogram ---
plt.figure(figsize=(6,4))
sns.histplot(data['Age_Band'], discrete=True, color="lightgreen")
plt.title("Histogram - Age Band")
plt.xlabel("Age_Band")
plt.show()

# --- Box Plot ---
plt.figure(figsize=(6,4))
sns.boxplot(x=data['Age_Band'], y=data['Avg_Daily_Screen_Time_hr'], palette="pastel")
plt.title("Boxplot - Screen Time by Age Band")
plt.show()

# --- Line Plot ---
plt.figure(figsize=(6,4))
sns.lineplot(x=data['Age_Band'].value_counts().index, y=data['Age_Band'].value_counts().values, marker="o", color="purple")
plt.title("Line Plot - Age Band Frequencies")
plt.xlabel("Age_Band")
plt.ylabel("Count")
plt.show()
```

```python
# ===================================================
# Primary Device
# ===================================================

# --- Bar Chart ---
plt.figure(figsize=(6,4))
data['Primary_Device'].value_counts().plot(kind='bar', color="skyblue")
plt.title("Bar Chart - Primary Device")
plt.xlabel("Device")
plt.ylabel("Count")
plt.show()

# --- Histogram ---
plt.figure(figsize=(6,4))
sns.histplot(data['Primary_Device'], discrete=True, color="lightgreen")
plt.title("Histogram - Primary Device")
plt.xlabel("Device")
plt.show()

# --- Box Plot ---
plt.figure(figsize=(6,4))
sns.boxplot(x=data['Primary_Device'], y=data['Avg_Daily_Screen_Time_hr'], palette="pastel")
plt.title("Boxplot - Screen Time by Device")
plt.show()

# --- Line Plot ---
plt.figure(figsize=(6,4))
sns.lineplot(x=data['Primary_Device'].value_counts().index, y=data['Primary_Device'].value_counts().values, marker="o", color="purple")
plt.title("Line Plot - Primary Device Frequencies")
plt.xlabel("Device")
plt.ylabel("Count")
plt.show()

```

```python
# ===================================================
#  Screen Time vs Gender
# ===================================================

# Bar Chart - Average Screen Time by Gender
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=data, palette="Blues")
plt.title("Bar Chart of Average Screen Time by Gender")
plt.show()

# Histogram - Screen Time Distribution split by Gender
plt.figure(figsize=(6,4))
sns.histplot(data=data, x="Avg_Daily_Screen_Time_hr", hue="Gender", bins=20, kde=True, palette="Set2")
plt.title("Histogram of Screen Time by Gender")
plt.show()

# Scatter Plot - Gender vs Screen Time
plt.figure(figsize=(6,4))
sns.stripplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=data, jitter=True, palette="Set1")
plt.title("Scatter Plot - Screen Time by Gender")
plt.show()

# Violin Plot - Screen Time distribution by Gender
plt.figure(figsize=(6,4))
sns.violinplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=data, palette="pastel")
plt.title("Violin Plot - Screen Time by Gender")
plt.show()
```

```python
# ===================================================
#  Screen Time vs Age Band
# ===================================================

# Bar Chart - Average Screen Time by Age Band
plt.figure(figsize=(6,4))
sns.barplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=data, palette="coolwarm")
plt.title("Bar Chart of Average Screen Time by Age Band")
plt.show()

# Histogram - Screen Time Distribution split by Age Band
plt.figure(figsize=(6,4))
sns.histplot(data=data, x="Avg_Daily_Screen_Time_hr", hue="Age_Band", bins=20, kde=True, palette="muted")
plt.title("Histogram of Screen Time by Age Band")
plt.show()

# Scatter Plot - Age Band vs Screen Time
plt.figure(figsize=(6,4))
sns.stripplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=data, jitter=True, palette="Set2")
plt.title("Scatter Plot - Screen Time by Age Band")
plt.show()

# Violin Plot - Screen Time distribution by Age Band
plt.figure(figsize=(6,4))
sns.violinplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=data, palette="pastel")
plt.title("Violin Plot - Screen Time by Age Band")
plt.show()
```

```python
# ===================================================
# Screen Time vs Urban_or_Rural (Location Type)
# ===================================================

# Bar Chart - Average Screen Time by Location
plt.figure(figsize=(6,4))
sns.barplot(x="Urban_or_Rural", y="Avg_Daily_Screen_Time_hr", data=data, palette="Greens")
plt.title("Bar Chart Average Screen Time by Location Type")
plt.show()

# Histogram - Screen Time Distribution split by Location
plt.figure(figsize=(6,4))
sns.histplot(data=data, x="Avg_Daily_Screen_Time_hr", hue="Urban_or_Rural", bins=20, kde=True, palette="Set1")
plt.title("Histogram of Screen Time by Location Type")
plt.show()

# Scatter Plot - Location vs Screen Time
plt.figure(figsize=(6,4))
sns.stripplot(x="Urban_or_Rural", y="Avg_Daily_Screen_Time_hr", data=data, jitter=True, palette="Set2")
plt.title("Scatter Plot - Screen Time by Location Type")
plt.show()

# Violin Plot - Screen Time distribution by Location
plt.figure(figsize=(6,4))
sns.violinplot(x="Urban_or_Rural", y="Avg_Daily_Screen_Time_hr", data=data, palette="pastel")
plt.title("Violin Plot - Screen Time by Location Type")
plt.show()
```

```python
# OBSERVATIONS

1. Most children spend between 4 to 6 hours on screens each day.

2. The '1115' age group has the highest number of students and also shows higher screen time.

3. Boys generally have slightly more screen time than girls.

4. Smartphones are the most commonly used devices among all students.

5. Urban children tend to have more screen time than rural children.

6. A few students have unusually high screen time compared to others (outliers).

7. Screen time gradually increases with age.

8. The number of boys and girls in the dataset is almost equal, which helps in fair analysis.

9. The spread of screen time is wider in older age groups, showing more variation.

10. Overall, screen habits seem to be influenced by age, gender, location, and access to devices.
```
