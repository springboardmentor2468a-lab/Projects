```python
# Load the dataset
import pandas as pd
df = pd.read_csv('Indian_Kids_Screen_Time.csv')
df.head()
```

```python
#Check Null values in the Dataset
df.isnull().sum()
```

```python
# Replace 'None' with 'No Health Impact'

df['Health_Impacts'] = df['Health_Impacts'].fillna('No Health Impact')
print("Data after replacing 'None':")
df.head()
```

```python
#Check Null values in the Dataset
df.isnull().sum()
```

```python
#Creating age bands

age_bins = [8, 12, 16, 18] 
age_labels = ['Pre-teen', 'Teenagers', 'Late-teens']
df['Age_Band'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=True, include_lowest=True)

print("\nData with Age Bands:")
df.head()
```

```python
#Categorize Health Impacts Column 
def categorize_health_impact(impact):
    physical_impacts = ['Poor Sleep', 'Eye Strain', 'Obesity Risk']
    mental_impacts = ['Anxiety']
    
    has_physical = any(p in impact for p in physical_impacts)
    has_mental = any(m in impact for m in mental_impacts)
    
    if has_physical and has_mental:
        return 'Physical, Mental Impact'
    elif has_physical:
        return 'Physical Impact'
    elif has_mental:
        return 'Mental Impact'
    else:
        return 'No Health Impact'

df['Health_Category'] = df['Health_Impacts'].apply(categorize_health_impact)

print("\nData with Health Category:")
print(df[['Health_Impacts', 'Health_Category']])
df.head(10)
```

```python
df.to_csv('Indian_Kids_Screen_Time_Updated.csv', index=False)
print("Dataset Modified")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset
df = pd.read_csv("Indian_Kids_Screen_Time_updated.csv")
df.head(5)
```

Univariate Analysis

```python
# Histogram for numerical variable
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=True, color='blue')
plt.title("Distribution of Daily Screen Time (hrs)")
plt.xlabel("Avg Daily Screen Time (hrs)")
plt.ylabel("Count")
plt.show()
```

#Insights 1.The majority of individuals have a daily screen time between 3 and 7 hours 2.A significant subset of the population reports very low screen time, specifically less than one hour per day.

```python
# Boxplot to detect outliers
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()
```

#Insights  1.The median age is approximately 13 years.2.The central 50% of the data  is concentrated between approximately 10 and 16 years, and the distribution is symmetrical.

```python
# Bar chart for categorical variable
sns.countplot(x='Urban_or_Rural', data=df)
plt.title("Urban vs Rural Distribution")
plt.show()
```

#Insights 1.The count of Urban observations is more than double the count of Rural observations.

```python
# Bar chart for Gender variable
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

#Insights 1.The dataset is nearly balanced in terms of gender distribution.There is a slight predominance of males in the dataset.

```python
# Age Band Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Age_Band", hue="Age_Band", data=df, palette="Set2", legend=False)
plt.title("Distribution of Kids by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Count of Kids")
plt.show()
```

#Insights The "Pre-teen" age band has the highest count of kids in the dataset, significantly outpacing the other two groups.

```python
# Exceeded Recommended Limit
sns.countplot(x='Exceeded_Recommended_Limit', hue ='Exceeded_Recommended_Limit', data=df, palette='Set2')
plt.title("Children Exceeding Recommended Screen Time")
plt.xlabel("Exceeded Limit")
plt.ylabel("Count")
plt.show()
```

#Insights 1.The vast majority of children in the dataset exceed the recommended screen time limit.

```python
# Distribution of Device Usage
plt.figure(figsize=(8,5))
sns.countplot(
    data = df,
    x='Primary_Device',
    hue='Primary_Device',
    order=df['Primary_Device'].value_counts().index,
    palette="Set3",
    legend=False
)
plt.title("Primary Device Usage")
plt.xlabel("Device Type")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()
```

#Insights 1.Laptops and Tablets are the least common primary devices, with similar, low counts.

```python
# Device Share Visual
device_counts = df['Primary_Device'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title("Primary Device Usage Distribution")
plt.show()
```

#Insights 1.Smartphones account for nearly half of all primary device usage, dominating the distribution.
          2.The above pie chart represents the primary usage distribution and also it can be clearly explained in the form of perrcentages like Smart            phones=47.05,Tablet=12.65,Laptop=14.8%,Tv=25.6%

Bivariate Analysis

```python
# Grouped bar plot - Age Band vs Exceeded Recommended Limit
sns.barplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title("Average Screen Time by Age Band")
plt.show()
```

#Insights 1.All age bands show a very high and similar average daily screen time, hovering around 4 to 4.5 hours.

```python
# Scatter Plot  Age vs Daily Screen Time
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, color='blue')
plt.title("Age vs Daily Screen Time")
plt.xlabel("Age")
plt.ylabel("Avg Daily Screen Time (hrs)")
plt.show()
```

#Insights 1.There is no clear linear correlation between age and average daily screen time, but high screen times are more common in younger ages (8-10).

```python
#Compare Screentime by Gender
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='Gender',
    y='Avg_Daily_Screen_Time_hr',
    hue='Gender',
    palette="coolwarm",
    legend=False
)
plt.title("Screentime by Gender")
plt.xlabel("Gender")
plt.ylabel("Daily Hours")
plt.show()
```

#Insights 1.The median daily screen time is nearly identical for both males and females, around 4.5 hours, indicating no significant central difference based on gender.

```python
# Heatmap  Average Screentime by Age Band and Gender
heatmap_data = df.pivot_table(
    index='Age_Band',
    columns='Gender',
    values='Avg_Daily_Screen_Time_hr',
    aggfunc='mean',
    observed=False  # removes FutureWarning
)
plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Average Screentime by Age Band and Gender")
plt.xlabel("Gender")
plt.ylabel("Age Band")
plt.show()
```

```python
# Line Plot - Avg_Daily_Screen_Time Vs Age_Band
plt.figure(figsize=(8,5))
sns.lineplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=df, marker='o', color='teal')
plt.title("Average Daily Screen Time by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Screen Time (hrs)")
plt.grid(True)
plt.show()
```

#Insights
1.This line plot shows the Avg daily screen time by age band
2.This pattern highlights that older kids spend significantly more time on devices
3.Teens (17-18 yrs) have the highest daily screen time
4.This plot shows the screen time increases with the increase of ages
5.Finally we can conclude that the increase of age will also increases the avg screen time as per the line plot.
