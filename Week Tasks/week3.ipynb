```python
import pandas as pd
df = pd.read_csv("Indian_Kids_Screen_Time.csv")
```

```python
# Univariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


# Numerical distributions


plt.figure(figsize=(6,4))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Daily Screen Time (hrs)")
plt.figtext(0.5, -0.05, "Most users spend around average daily screen time between 3-6 hrs.", ha="center", fontsize=9)
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=15, kde=True, color='orange')
plt.title("Distribution of Age")
plt.figtext(0.5, -0.05, "The dataset covers a wide age range, with more participants in young adult groups.", ha="center", fontsize=9)
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['Educational_to_Recreational_Ratio'], bins=20, kde=True, color='green')
plt.title("Distribution of Educational to Recreational Ratio")
plt.figtext(0.5, -0.05, "The ratio shows how users balance educational vs recreational screen usage.", ha="center", fontsize=9)
plt.show()

# Categorical distributions


plt.figure(figsize=(6,4))
sns.countplot(x='Primary_Device', data=df)
plt.title("Distribution of Primary Device Usage")
plt.xticks(rotation=45)
plt.figtext(0.5, -0.05, "Smartphones and laptops appear as the most frequently used devices.", ha="center", fontsize=9)
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.figtext(0.5, -0.05, "The dataset contains both male and female users with slight variations in counts.", ha="center", fontsize=9)
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Urban_or_Rural', data=df)
plt.title("Urban vs Rural Distribution")
plt.figtext(0.5, -0.05, "Urban users are more represented compared to rural users.", ha="center", fontsize=9)
plt.show()

# Pie Charts

# Gender Pie Chart
plt.figure(figsize=(6,6))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'], explode=[0.05,0.05])
plt.title('Gender Distribution (Pie Chart)')
plt.figtext(0.5, -0.05, "The pie chart shows the proportion of male and female users.", ha="center", fontsize=9)
plt.show()

# Primary Device Pie Chart
plt.figure(figsize=(6,6))
device_counts = df['Primary_Device'].value_counts()
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=90, explode=[0.05]*len(device_counts))
plt.title('Primary Device Distribution (Pie Chart)')
plt.figtext(0.5, -0.05, "Smartphones dominate as the primary device for screen usage.", ha="center", fontsize=9)
plt.show()

# Urban vs Rural Pie Chart
plt.figure(figsize=(6,6))
location_counts = df['Urban_or_Rural'].value_counts()
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', startangle=90, colors=['orange','lightblue'], explode=[0.05,0.05])
plt.title('Urban vs Rural Distribution (Pie Chart)')
plt.figtext(0.5, -0.05, "Urban users form the majority, with fewer rural participants.", ha="center", fontsize=9)
plt.show()
```

```python
# Bivariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 1. Screen Time by Gender (Boxplot)
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title("Screen Time by Gender")
plt.figtext(0.5, -0.05, "Boxplot shows variation in screen time between male and female users.", ha="center", fontsize=9)
plt.show()

# 2. Screen Time by Urban vs Rural (Barplot)
plt.figure(figsize=(6,4))
sns.barplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean')
plt.title("Average Screen Time by Location")
plt.figtext(0.5, -0.05, "Urban users generally spend more screen time than rural users.", ha="center", fontsize=9)
plt.show()

# 3. Screen Time vs Exceeded Recommended Limit (Barplot)
plt.figure(figsize=(6,4))
sns.barplot(x='Exceeded_Recommended_Limit', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean')
plt.title("Screen Time vs Exceeding Recommended Limit")
plt.figtext(0.5, -0.05, "Users who exceed the recommended screen limit have higher average screen time.", ha="center", fontsize=9)
plt.show()

# 4. Age vs Screen Time (Scatter by Gender)
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', hue='Gender', data=df)
plt.title("Age vs Screen Time by Gender")
plt.figtext(0.5, -0.05, "Scatter plot shows how screen time varies with age, differentiated by gender.", ha="center", fontsize=9)
plt.show()

# 5. Age vs Screen Time (Line Plot)
plt.figure(figsize=(6,4))
sns.lineplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, marker="o", color="green")
plt.title("Screen Time Trend with Age")
plt.figtext(0.5, -0.05, "Line plot shows the trend of average screen time increasing or decreasing with age.", ha="center", fontsize=9)
plt.show()

# 6. Screen Time by Primary Device (Barplot)
plt.figure(figsize=(7,4))
sns.barplot(x='Primary_Device', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean')
plt.title("Average Screen Time by Primary Device")
plt.xticks(rotation=45)
plt.figtext(0.5, -0.05, "Smartphone users tend to have higher average screen time compared to other devices.", ha="center", fontsize=9)
plt.show()

# 7. Educational to Recreational Ratio vs Screen Time (Scatter)
plt.figure(figsize=(6,4))
sns.scatterplot(x='Educational_to_Recreational_Ratio', y='Avg_Daily_Screen_Time_hr', data=df, hue='Gender')
plt.title("Educational vs Recreational Ratio and Screen Time")
plt.figtext(0.5, -0.05, "Scatter shows whether a balanced ratio affects total daily screen time.", ha="center", fontsize=9)
plt.show()

# 8. Screen Time by Gender and Location (Grouped Barplot)
plt.figure(figsize=(6,4))
sns.barplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', hue='Gender', data=df, estimator='mean')
plt.title("Average Screen Time by Gender and Location")
plt.figtext(0.5, -0.05, "Urban males and females show slightly higher screen times compared to rural counterparts.", ha="center", fontsize=9)
plt.show()

# 9. Correlation Heatmap (Numerical Variables)
plt.figure(figsize=(6,4))
sns.heatmap(df[['Age','Avg_Daily_Screen_Time_hr','Educational_to_Recreational_Ratio']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Age, Screen Time & Edu/Recreation Ratio")
plt.figtext(0.5, -0.05, "Heatmap shows correlation: age may have mild effect, while ratio impacts screen hours differently.", ha="center", fontsize=9)
plt.show()
```
