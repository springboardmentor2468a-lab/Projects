```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style="whitegrid")

# 1. Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True, color="skyblue")
plt.title("Age Distribution of Kids")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()



```

```python
# 2. Average Daily Screen Time
plt.figure(figsize=(8,5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], bins=20, kde=True, color="salmon")
plt.title("Daily Screen Time Distribution")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Number of Kids")
plt.show()
```

```python
# 3. Gender Distribution
plt.figure(figsize=(6,5))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# 4. Primary Device Used
plt.figure(figsize=(7,5))
sns.countplot(x="Primary_Device", data=df,  order=df['Primary_Device'].value_counts().index)
plt.title("Most Common Devices for Screen Time")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Exceeded Recommended Limit
plt.figure(figsize=(6,5))
sns.countplot(x="Exceeded_Recommended_Limit", data=df)
plt.title("Exceeded Recommended Screen Time?")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y="Health_Impacts",

    data=df,
    order=df['Health_Impacts'].value_counts().head(10).index,

    legend=False
)
plt.title("Top Reported Health Impacts")
plt.show()

```

```python

# 7. Urban vs Rural
plt.figure(figsize=(6,5))
sns.countplot(x="Urban_or_Rural", data=df,)
plt.title("Urban vs Rural Kids")
plt.show()
```

```python
#heat map
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(),
           annot=True,  cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

After looking at the dataset, here is what I haveve understood:

Structure of the data
The dataset includes information like Age, Gender, how much time kids spend on screens daily, the type of device they use, and other details such as whether they exceeded recommended screen limits and their educational-to-recreational activity ratio. It also has location information like Urban or Rural areas.

Missing Data
 Health Impacts column. This column has a large number of missing values (3218 entries missing), which is something we need to explore further.



1. Age Distribution

Kids are between 8 and 18 years old.

Most students fall in the 12 to 16 age range.
2. Daily Screen Time

Some kids spend up to 14 hours/day on screens.

The majority are above the recommended healthy limit

3. Gender Distribution

Slightly more males (4942) than females (4770)

Both groups have similar screen time patterns.

 4. Primary Device Used

Smartphones dominate

Other devices: Tablets, Laptops, and TVs, but far less frequent.

5. Exceeded Recommended Screen Time

85% of kids (8301 out of 9712) exceed safe daily screen time.

Only a small portion stays within healthy limits.
6. Health Impacts

Poor Sleep (highest)

Eye Strain

Headaches

7. Urban vs Rural Distribution

More kids are from Urban areas (≈ 70%) than Rural.
