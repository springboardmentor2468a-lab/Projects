```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual theme
sns.set(style="whitegrid", palette="Set2")
df = pd.read_csv("Indian_Kids_Screen_Time.csv")
```

```python
#1.device mix by gender
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', hue='Primary_Device')
plt.title("Device Mix by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.legend(title="Device Type")
plt.show()
```

1.Males and females show different preferences for digital devices.

2.Smartphones dominate for both genders, but laptop usage tends to be slightly higher among males.

3.Females use mobile devices more frequently for multitasking and communication.

4.Overall, gender differences indicate varying digital habits and access priorities.

```python
#2.device mix by age
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Age_Group', hue='Primary_Device')
plt.title("Device Mix by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count of Users")
plt.legend(title="Device Type")
plt.show()


```

1.Younger users (1825 years) prefer smartphones due to portability and entertainment.

2.Middle-aged users (2640 years) balance between laptops and mobiles, likely for work and study.

3.Older users (40+ years) favor larger devices like TVs for easier viewing.

4.Device preference clearly shifts from mobile-first to larger screens with age.

```python
#3.Device Mix by Urban vs Rural
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Urban_or_Rural', hue='Primary_Device')
plt.title("Device Mix by Urban vs Rural Users")
plt.xlabel("Location Type")
plt.ylabel("Count of Users")
plt.legend(title="Device Type")
plt.show()
```

1.Urban users show greater diversity in device usage, especially smartphones and laptops.

2.Rural users rely more on basic mobile phones and shared devices.

3.The difference highlights a digital divide in access and affordability.

4.Urban populations benefit from better internet connectivity and technology exposure.

```python
#4.Activity Type by Gender
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', hue='Primary_Device')
plt.title("Activity Type by Gender")
plt.xlabel("Gender")
plt.ylabel("Count of Users")
plt.legend(title="Primary_Device")
plt.show()
```

1.Males engage more in recreational activities such as gaming and streaming.

2.Females spend comparatively more time on educational or productive tasks.

3.The pattern shows gender-based variation in digital engagement purposes.

4.Recreational activity dominates overall screen use across both genders.

```python
#5.Average Screen Time (hours)
plt.figure(figsize=(7,4))
sns.barplot(data=df, x='Age_Group', y='Avg_Daily_Screen_Time_hr', hue='Gender', estimator='mean')
plt.title("c")
plt.xlabel("Age Group")
plt.ylabel("Average Screen Time (hours)")
plt.legend(title="Gender")
plt.show()
```

1.Younger users (1825) record the highest daily screen time, showing heavy digital engagement.

2.Males generally spend more time on screens than females in each age group.

3.Screen time gradually declines with age, reflecting reduced digital dependency.



```python
#6.Overall Device Preference Distribution
plt.figure(figsize=(5,5))
device_counts = df['Primary_Device'].value_counts()
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Overall Device Preference Distribution")
plt.show()
```

1.Smartphones dominate overall device usage across all demographics.

2.Laptops and tablets follow as secondary devices for productivity and education.

3.Larger devices like TVs and desktops hold a smaller usage share.

4.This reinforces a mobile-first trend in digital behavior.

```python
#1.Activity Type by Age Group (Heatmap)
heat_data = df.groupby(['Age_Group', 'Primary_Device']).size().unstack(fill_value=0)
plt.figure(figsize=(6,4))
sns.heatmap(heat_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Activity Category Distribution by Age Group")
plt.xlabel("Primary_Device")
plt.ylabel("Age Group")
plt.show()
```

1.Younger age groups show high recreational activity, reflecting social media and entertainment trends.

2.Older users focus more on educational or work-related digital content.

3.The trend shows an age-dependent shift in screen usage priorities.

4.Overall, recreational usage peaks among youth, while education dominates older age brackets.

```python
print(df.columns)
# Define bins and labels
bins = [7, 12, 15, 18]
labels = ["Pre-Teens (7-12)", "Teenagers (13-15)", "Late Teens (16-18)"]

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

print(df[['Age', 'Age_Group']])
```

```python
#2.Correlation Between Key Numerical Features
plt.figure(figsize=(7,5))
num_cols = ['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Key Numerical Features")
plt.show()
```

```python
#3.Device Usage by Gender and Urban/Rural Location
heat_data2 = df.groupby(['Gender', 'Urban_or_Rural'])['Primary_Device'].value_counts().unstack(fill_value=0)
plt.figure(figsize=(7,5))
sns.heatmap(heat_data2, annot=True, cmap='OrRd', fmt='d')
plt.title("Device Usage by Gender and Urban/Rural Location")
plt.xlabel("Primary Device")
plt.ylabel("Gender & Location")
plt.show()

```

1.Reveals how device choices vary between urban/rural males and females.

2.Urban males tend to use laptops and smartphones more frequently.

3.Rural females show higher reliance on mobile devices for daily use.

4,Demonstrates how gender and geography combine to shape digital behavior.

```python
#4.Average Daily Screen Time by Gender and Age Group
heat_data3 = df.groupby(['Gender', 'Age_Group'])['Avg_Daily_Screen_Time_hr'].mean().unstack()
plt.figure(figsize=(7,5))
sns.heatmap(heat_data3, annot=True, cmap='Purples', fmt=".2f")
plt.title("Average Daily Screen Time by Gender and Age Group")
plt.xlabel("Age Group")
plt.ylabel("Gender")
plt.show()
```

1.Identifies which genderage combinations show the highest average screen time.

2.Younger males (1825) record the peak daily screen usage.

3.Screen time declines steadily with age across both genders.

4.Useful for recognizing high-engagement cohorts for digital targeting.

```python
#5.Exceeded Recommended Screen Time by Age and Gender
heat_data4 = df.groupby(['Age_Group', 'Gender'])['Exceeded_Recommended_Limit'].value_counts().unstack(fill_value=0)
plt.figure(figsize=(7,5))
sns.heatmap(heat_data4, annot=True, cmap='coolwarm', fmt='d')
plt.title("Exceeded Recommended Screen Time by Age and Gender")
plt.xlabel("Exceeded Recommended Limit (0=No, 1=Yes)")
plt.ylabel("Age Group & Gender")
plt.show()
```

1.Displays which age and gender groups are most likely to exceed the safe screen time limit.

2.Young males (1825) show the highest frequency of exceeding recommended usage.

3.Older females (40+) generally stay within healthy limits.

4.Reveals age-related self-regulation and digital overuse risks.

```python
#6.Educational vs Recreational Usage by Gender and Location
heat_data5 = df.groupby(['Gender', 'Urban_or_Rural'])['Educational_to_Recreational_Ratio'].mean().unstack()
plt.figure(figsize=(7,5))
sns.heatmap(heat_data5, annot=True, cmap='Greens', fmt=".2f")
plt.title("Educational vs Recreational Usage by Gender and Location")
plt.xlabel("Location Type")
plt.ylabel("Gender")
plt.show()
```

1.Shows how educational vs. recreational usage differs across gender and geography.

2.Urban females tend to have a higher educational-to-recreational ratio, suggesting more purposeful usage.

3.Rural males exhibit higher recreational activity levels.

4.Provides insight into learning-oriented vs leisure-oriented behavior in different groups.
