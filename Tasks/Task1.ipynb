```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style="whitegrid")

# 1. Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True, color="skyblue")
plt.title("Age Distribution of Kids")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Average Daily Screen Time
plt.figure(figsize=(8,5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], bins=20, kde=True, color="salmon")
plt.title("Daily Screen Time Distribution")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Number of Kids")
plt.show()
```

```python
# 3. Gender Distribution
plt.figure(figsize=(6,5))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()

```

```python

plt.figure(figsize=(7,5))
sns.countplot(x="Device_Most_Used", data=df,  order=df['Device_Most_Used'].value_counts().index)
plt.title("Most Common Devices for Screen Time")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(6,5))
sns.countplot(x="Parental_Control", data=df)
plt.title("Parental Control Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y="Device_Most_Used",
    data=df,
    order=df['Device_Most_Used'].value_counts().head(10).index,
    legend=False
)
plt.title("Top Reported Device_Most_Used")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x="Weekday_ScreenTime_Hours", data=df)
plt.title("Weekday Screen Time Distribution")
plt.show()
```
