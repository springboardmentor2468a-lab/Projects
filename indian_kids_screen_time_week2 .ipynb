# WEEK 2

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Replace 'None' with "No Health Impacts"
df['Health_Impacts'] = df['Health_Impacts'].replace('None', "No Health Impacts").fillna("No Health Impacts")
df[['Age','Avg_Daily_Screen_Time_hr','Exceeded_Recommended_Limit','Health_Impacts']].head(10)
```

```python
#Age Bands
def life_stage(age):
    if 8 <= age <= 12:
        return "Pre-Teens"
    elif 13 <= age <= 16:
        return "Teenagers"
    elif 17 <= age <= 18:
        return "Late Teens"

df['Life_Stage'] = df['Age'].apply(life_stage)
# Show first 20 rows with Age and Life_Stage
df[['Age', 'Life_Stage']].head(5)
```

```python
# Derived feild for Devices
def device_type(device):
    if device in ["Smartphone", "Tablet", "Laptop"]:
        return "Portable"
    elif device == "TV":
        return "Fixed"
    else:
        return "Other"

df['Device_Type'] = df['Primary_Device'].apply(device_type)
df[['Primary_Device', 'Device_Type']].head(7)
```

```python
#Derived feild for health_impacts
def health_status(impact):
    impact = str(impact).lower()  # ensure string
    if "no health impacts" in impact:
        return "Healthy"
    
    mental = ["anxiety", "poor sleep"]
    physical = ["eye strain", "obesity risk"]
    
    has_mental = any(m in impact for m in mental)
    has_physical = any(p in impact for p in physical)
    
    if has_mental and has_physical:
        return "Mental, Physical"
    elif has_mental:
        return "Mental"
    elif has_physical:
        return "Physical"

df['Health_Status'] = df['Health_Impacts'].apply(health_status)
df[['Health_Impacts','Health_Status']].head(6)
```

```python
# Save as CSV
df.to_csv("preprocessed_kids_screen_time.csv", index=False)
print("Dataset saved successfully!")
```
