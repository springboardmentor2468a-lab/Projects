```python
# Replace 'None' with "No Health Impacts"
df['Health_Impacts'] = df['Health_Impacts'].fillna('No health impacts')
df[['Age','Avg_Daily_Screen_Time_hr','Exceeded_Recommended_Limit','Health_Impacts']].head(10)
```

```python
#Check for missing (null)
df.isnull().sum()
```

```python
# Categorising Age bands
age_bins = [8, 12, 16, 18] 
age_labels = ['Pre-teen', 'Teenagers', 'Late-teens']

df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=True, include_lowest=True)
df[['Age', 'Age_Group']].head(5)
```

```python
# Device type: Fixed or Portable
def device_type(device):
    if device in ["Smartphone", "Tablet", "Laptop"]:
        return "Portable"
    else:
        return "Fixed"

df['Device_Type'] = df['Primary_Device'].apply(device_type)
df[['Primary_Device', 'Device_Type']].head(7)
```

```python
# Screen size: TV >=30", ['Smartphone', 'Laptop', 'Tablet'] <30"
def screen_size(device):
    if device == 'TV':
        return '>=30'
    else:
        return '<30' 
df['Screen_Size'] = df['Primary_Device'].apply(screen_size)
df[['Primary_Device', 'Device_Type', 'Screen_Size']].head(7)
```

```python
# Categorize health impacts: Mental, Physical, Both, or No Health Impacts
def health_status(impact):
    impact = str(impact).lower()
    
    if "no health impacts" in impact:
        return "No Health Impacts"   
    
    mental = ["anxiety", "poor sleep"]
    physical = ["eye strain", "obesity risk"]
    
    has_mental = any(m in impact for m in mental)
    has_physical = any(p in impact for p in physical)
    
    if has_mental and has_physical:
        return "Mental, Physical"
    elif has_mental:
        return "Mental"
    elif has_physical:
        return "Physical"

df['Health_Status'] = df['Health_Impacts'].apply(health_status)
df[['Health_Impacts','Health_Status']].head(14)
```

```python
# Reorder columns
df = df[['Age',
         'Age_group',
         'Gender',
         'Avg_Daily_Screen_Time_hr',
         'Primary_Device',
         'Device_Type',
         'Screen_Size',
         'Exceeded_Recommended_Limit',
         'Educational_to_Recreational_Ratio',
         'Health_Impacts',
         'Health_Status',
         'Urban_or_Rural']]
```

```python
# Save as CSV
df.to_csv("preprocessed_kids_screen_time.csv", index=False)
print("Dataset saved successfully!")
```
