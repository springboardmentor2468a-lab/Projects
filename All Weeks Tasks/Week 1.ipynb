```python
import seaborn as sns
import matplotlib.pyplot as plt

print("Correlation matrix")
correlation = df.corr(numeric_only=True) 
print(correlation)

sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation of all numerical columns")
plt.show()
```

```python
Q1 = df['Avg_Daily_Screen_Time_hr'].quantile(0.25)
Q3 = df['Avg_Daily_Screen_Time_hr'].quantile(0.75)
IQR = Q3 - Q1
print("IQR:", IQR)
```

```python
screen_time = (
    df.groupby('Urban_or_Rural', as_index=False)['Avg_Daily_Screen_Time_hr']
      .mean()
      .round(2)
      .rename(columns={'Avg_Daily_Screen_Time_hr': 'Average_Screen_Time'})
)

print(screen_time)
```
