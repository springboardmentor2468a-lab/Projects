```python
# Create Age Bands
bins = [7, 12, 16, 18]
labels = ["Pre-teens (8-12)", "Teenagers (13-16)", "Late Teens (17-18)"]
df["Age_Band"] = pd.cut(df["Age"], bins=bins, labels=labels, include_lowest=True)

print(df[["Age", "Age_Band"]].head(15))

import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing Age Band distribution

# Set style
sns.set(style="whitegrid")

# Plot Age Band distribution
plt.figure(figsize=(8, 5))
sns.countplot(x="Age_Band", data=df, palette="pastel")

plt.title("Age Band Distribution of Indian Kids", fontsize=14)
plt.xlabel("Age Band", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()

```

```python

from pandas.api.types import CategoricalDtype   # <-- FIX

# Device Categorization by Screen Size

device_screen_size = {
    "Mobile": "<30",
    "Tablet": "<30",
    "Laptop": "<30",
    "Desktop": ">=30",
    "Television": ">=30"
}

df["Device_Category"] = df["Primary_Device"].map(device_screen_size)

# Force category order so both <30 and >=30 always appear
cat_type = CategoricalDtype(categories=["<30", ">=30"], ordered=True)
df["Device_Category"] = df["Device_Category"].astype(cat_type)

# Screen Time Categorization

df["Screen_Time_Category"] = pd.cut(
    df["Avg_Daily_Screen_Time_hr"],
    bins=[0, 2, 4, df["Avg_Daily_Screen_Time_hr"].max()],
    labels=["Low (0-2 hrs)", "Medium (2-4 hrs)", "High (>4 hrs)"],
    include_lowest=True
)


# Visualization

plt.figure(figsize=(12,5))

# Plot 1: Device Category Count
plt.subplot(1, 2, 1)
sns.countplot(x="Device_Category", data=df, palette="Set2")
plt.title("Device Categorization by Screen Size")
plt.xlabel("Device Category")
plt.ylabel("Count")

# Plot 2: Screen Time Distribution
plt.subplot(1, 2, 2)
sns.countplot(x="Screen_Time_Category", data=df, palette="Set1")
plt.title("Screen Time Categorization")
plt.xlabel("Screen Time Category")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Standardize Health_Impacts column
df['Health_Impacts'] = df['Health_Impacts'].astype(str).str.strip().str.lower()

df['Health_Impacts'].replace(
    ["none", "no health impacts", "nan", "na", "null", ""],
    "No Health Impacts",
    inplace=True
)

print(df['Health_Impacts'].value_counts())
```

```python
# Define categories
physical_terms = ["eye strain", "headache", "obesity", "back pain", "sleep issues"]
mental_terms = ["anxiety", "depression", "stress", "attention problems"]

def categorize_health(impact):
    if impact in physical_terms:
        return "Physical"
    elif impact in mental_terms:
        return "Mental"
    elif impact == "No Health Impacts":
        return "None"
    else:
        return "Other"

df["Health_Category"] = df["Health_Impacts"].apply(categorize_health)

print(df[["Health_Impacts", "Health_Category"]].head(20))
```

```python
# Group by Age_Band and Health_Category
summary = df.groupby(["Age_Band", "Health_Category"]).size().unstack(fill_value=0)

print(" Health Impact Distribution by Age Band:")
print(summary)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("clean_indian_kids_screen_time.csv")

print("✅ Data Loaded")
print(df.head())
print(df.info())

# Set a style
sns.set(style="whitegrid")

# =======================
# 📊 Univariate Analysis
# =======================

# 1. Distribution of Daily Screen Time
plt.figure(figsize=(6,4))
sns.histplot(df["Avg_Daily_Screen_Time_hr"], bins=10, kde=True)
plt.title("Distribution of Daily Screen Time (hrs)")
plt.show()

# 2. Age Band distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Age_Band", data=df, order=df["Age_Band"].value_counts().index)
plt.title("Distribution of Age Bands")
plt.show()

# 3. Device Category distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Device_Category", data=df, order=["<30", ">30"])
plt.title("Device Category Distribution")
plt.show()

# 4. Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()

# 5. Location distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Urban_or_Rural", data=df)
plt.title("Urban vs Rural Distribution")
plt.show()
```

```python
# =======================
# 📊 Bivariate Analysis
# =======================

# 1. Screen time by Gender
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=df)
plt.title("Screen Time by Gender")
plt.show()

# 2. Screen time by Age Band
plt.figure(figsize=(6,4))
sns.boxplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=df,
            order=["Pre-teens (9-12)", "Teenagers (13-16)", "Late Teens (17-18)"])
plt.title("Screen Time by Age Band")
plt.show()

# 3. Screen time by Urban/Rural
plt.figure(figsize=(6,4))
sns.barplot(x="Urban_or_Rural", y="Avg_Daily_Screen_Time_hr", data=df, estimator="mean")
plt.title("Average Screen Time by Location Type")
plt.show()

# 4. Device Category vs Gender
plt.figure(figsize=(6,4))
sns.countplot(x="Device_Category", hue="Gender", data=df, order=["<30", ">30"])
plt.title("Device Category by Gender")
plt.show()

# 5. Age Band vs Location (Stacked Bar)
pd.crosstab(df["Age_Band"], df["Urban_or_Rural"]).plot(kind="bar", stacked=True, figsize=(7,5))
plt.title("Age Band vs Urban/Rural Distribution")
plt.ylabel("Count")
plt.show()
```

```python
# =======================
# 📊 Extra Insights
# =======================

# 1. Health Impacts frequency
plt.figure(figsize=(8,4))
sns.countplot(y="Health_Impacts", data=df, order=df["Health_Impacts"].value_counts().index)
plt.title("Reported Health Impacts")
plt.show()

# 2. Age Band vs Screen Time Category (Heatmap)
ct = pd.crosstab(df["Age_Band"], df["Screen_Time_Category"])
sns.heatmap(ct, annot=True, fmt="d", cmap="Blues")
plt.title("Age Band vs Screen Time Category")
plt.show()

# 3. Line plot: Screen Time trend with Age
plt.figure(figsize=(6,4))
sns.lineplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, marker="o")
plt.title("Trend of Screen Time with Age")
plt.show()

```

##  Univariate Analysis (Single Variable Insights)

###Age Bands

###Device Type vs Screen Time
