```python
#Finding Interquartile Range for all numerical Column
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

def calculate_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return pd.DataFrame({"Q1": Q1, "Q3": Q3, "IQR": IQR})


iqr_result = calculate_iqr(numeric_cols)

print("Interquartile Range (IQR) for all numeric columns:\n")
print(iqr_result)
```

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
#Loading the dataset
df = pd.read_csv("Indian_Kids_Screen_Time.CSV")
df.head(10)
```

## Handle Missing Values and Inconsistent Categories

```python
# Check missing values
print("Missing values before handling:\n", df.isnull().sum(), "\n")
```

```python
# Option 1: Fill missing numerical values with mean
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Option 2: Handling missing categorical values 
# Replace NaN/null in Health_Impacts with "No health impact"
df["Health_Impacts"] = df["Health_Impacts"].fillna("No health impact")

# Verify changes
print(df["Health_Impacts"].isnull().sum())  # should be 0 now
print(df["Health_Impacts"].value_counts().head())
```

```python
# Optional: Verify changes
print("After cleaning:\n", df.head())
```

```python
print("Missing values after handling:\n", df.isnull().sum(), "\n")
```

## Create Derived Field

```python
#Create Age Bands
bins = [8, 12, 16, 18]  # boundaries
labels = ["Pre-teens", "Teenagers", "Late teens"]

# Create new column Age_Band
df["Age_Band"] = pd.cut(df["Age"], bins=bins, labels=labels, include_lowest=True)

# Verify result
print(df[["Age", "Age_Band"]].head(10))
```

## Categorization of Health impact column

```python
# Categorization of Health_Impacts column into Health_Impact_Category Column
physical = ["Eye Strain","Headache","Poor Posture","Obesity","Fatigue"]
mental = ["Poor Sleep","Anxiety","Stress","Depression","Addiction"]

df["Health_Impact_Category"] = df["Health_Impacts"].apply(
    lambda x: "No Impact" if x=="None" 
    else "Physical&Mental" if any(p in x for p in physical) and any(m in x for m in mental)
    else "Physical" if any(p in x for p in physical)
    else "Mental" if any(m in x for m in mental)
    else "No Impact"
)
df.head(10)
```

## Categorization of Primary_Device Column

```python
# Categorize Primary_Device column
df["Device_Category"] = df["Primary_Device"].apply(
    lambda x: "Wallmounted device" if x == "TV" else "Portable device"
)

# Verify result
print(df[["Primary_Device", "Device_Category"]].head(10))
print(df["Device_Category"].value_counts())
```

## creating derived fields : age bands, weekday/weekend flags, device/activity shares , formating any date/time fields

```python
# Get column names as a list
column_names_list = df.columns.tolist()
print("\nColumn names (list):")
column_names_list
```

```python
#Device Share (Proportion of each device used)
# Count device usage and calculate share
device_share = df["Primary_Device"].value_counts(normalize=True) * 100
print("Device Share(% of kids using each device):\n", device_share, "\n")
```

```python
#Educational vs Recreational Share
df["Educational_Share"] = df["Educational_to_Recreational_Ratio"]
df["Recreational_Share"] = 1 - df["Educational_to_Recreational_Ratio"]
Educational = df["Educational_Share"].mean() * 100
Recreational = df["Recreational_Share"].mean() * 100
# avg_activity_share = {
#     "Educational": df["Educational_Share"].mean() * 100,
#     "Recreational": df["Recreational_Share"].mean() * 100
# }
print("Activity Share(% of kids whose spending of time):\n")
print("Educational : ", Educational)
print("Recreational :", Recreational)
```

## Save Preprocessed data for reuse

```python
# Save Preprocessed Dataset
df.to_csv("Indian_Kids_Screen_Time_Updated.csv", index=False)

print("Preprocessed dataset saved successfully!")
```

```python
# Insights
1. A large portion exceeds the recommended screen time limits.
2. Indicates growing risk of digital overexposure, especially among teens.
3. Useful for framing health-related interventions.
```

```python
# Device Share Visual
device_counts = df['Primary_Device'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title("Primary Device Usage Distribution")
plt.show()
```

```python
# Insights
1. In Percentage most of the Kids uses Smartphone which is 47%, and Tv 25.6%, Laptop 14.8% and Tablet 12.6%.
```

```python
#Educational vs Recreational Share
df["Educational_Share"] = df["Educational_to_Recreational_Ratio"]
df["Recreational_Share"] = 1 - df["Educational_to_Recreational_Ratio"]
avg_activity_share = {
    "Educational": df["Educational_Share"].mean() * 100,
    "Recreational": df["Recreational_Share"].mean() * 100
}
plt.figure(figsize=(6,6))
plt.pie(avg_activity_share.values(), labels=avg_activity_share.keys(),
        autopct="%.1f%%", startangle=90, colors=["#4CAF50", "#FF7043"])
plt.title("Average Screen Time Split: Educational vs Recreational")
plt.show()
```

```python
# Plot pie chart
# Count the number of occurrences for each device category
device_counts = df['Device_Category'].value_counts()

plt.figure(figsize=(6,5))
plt.pie(
    device_counts.values,
    labels=device_counts.index,
    autopct='%.1f%%',      # show percentages
    startangle=90,         # start from top
    colors=['lightblue', '#1f77b4'],  # match your uploaded image style
)
plt.title('Device Category Distribution')
plt.axis('equal')          # makes the pie chart a circle
plt.show()
```

```python
# Insight Summary:
1. Portable devices make up ~74% of total usage, dominating kids' screen time.
2. Wallmounted devices (e.g., TVs) account for around 26%, indicating shared, family usage.
3. This highlights that personal, portable screens are the primary source of screen exposure for children.
```

```python
 # Plot pie chart
health_counts = df['Health_Impact_Category'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(
        health_counts.values,
        labels=health_counts.index,
        autopct='%.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors
    )
plt.title("Health Impact Category Distribution")
plt.axis('equal')  # Makes the pie a perfect circle
plt.show()
```

```python
#Insight Summary:
1. The charts show the proportion and frequency of different health impact categories caused by screen time.
2. Typically, categories like â€œModerate Impactâ€ or â€œHigh Impactâ€ dominate, indicating widespread health concerns (e.g., sleep issues, eye strain).
```
