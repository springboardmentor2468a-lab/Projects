## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
#Load the dataset
df = pd.read_csv("Indian_Kids_Screen_Time_Updated.csv")
df.head(10)
```

## Segment-wise Summary

```python
# 1. Age Band × Gender × Device Category
summary1 = df.groupby(["Age_Band", "Gender", "Device_Category"])["Avg_Daily_Screen_Time_hr"] \
             .mean().reset_index().rename(columns={"Avg_Daily_Screen_Time_hr":"Average_Screen_Time_hr"})
print("\n1️⃣ Summary: Age Band vs Gender vs Device Category")
display(summary1)

# 2. Age Band × Screen Time Category
summary2 = df.groupby(["Age_Band", "Device_Category"])["Avg_Daily_Screen_Time_hr"] \
             .mean().reset_index().rename(columns={"Avg_Daily_Screen_Time_hr":"Average_Screen_Time_hr"})
print("\n2️⃣ Summary: Age Band vs Device Category")
display(summary2)

# 3. Gender × Device Category
summary3 = df.groupby(["Gender", "Device_Category"])["Avg_Daily_Screen_Time_hr"] \
             .mean().reset_index().rename(columns={"Avg_Daily_Screen_Time_hr":"Average_Screen_Time_hr"})
print("\n3️⃣ Summary: Gender vs Device Category")
display(summary3)

# 4. Urban vs Rural × Device Category
summary4 = df.groupby(["Urban_or_Rural", "Device_Category"])["Avg_Daily_Screen_Time_hr"] \
             .mean().reset_index().rename(columns={"Avg_Daily_Screen_Time_hr":"Average_Screen_Time_hr"})
print("\n4️⃣ Summary: Urban vs Rural vs Device Category")
display(summary4)

# 5. Health Category × Age Band
summary5 = df.groupby(["Health_Impact_Category", "Age_Band"])["Avg_Daily_Screen_Time_hr"] \
             .mean().reset_index().rename(columns={"Avg_Daily_Screen_Time_hr":"Average_Screen_Time_hr"})
print("\n5️⃣ Summary: Health Impact Category vs Age Band")
display(summary5)

# 6. Exceeded Recommended Limit by Age Band
summary6 = df.groupby(["Age_Band", "Exceeded_Recommended_Limit"]).size().reset_index(name="Count")
print("\n 6 Summary: Age Band vs Exceeded Recommended Limit")
display(summary6)

# 7. Education vs Recreation Ratio by Device Type
summary7 = df.groupby("Device_Category")["Educational_to_Recreational_Ratio"] \
             .mean().reset_index().rename(columns={"Educational_to_Recreational_Ratio":"Avg_EdToRec_Ratio"})
print("\n 7 Summary: Device Type vs Educational to Recreational Ratio")
display(summary7)
```

```python
# Insights :
1. Teenagers have the highest average screen time compared to other age groups.
2. Smartphones are the most used device across all age groups.
3. Male users show slightly higher screen usage than females in most device categories.
4. Portable devices dominate usage, especially among Pre-teens and Teenagers.
5. High screen time cases are most common among Teenagers, followed by Pre-teens.
6. Late-teens mostly fall into Medium screen time usage.
7. Males use smartphones slightly more than females.
8. Females use laptops slightly more, possibly due to academic purposes.
9. TV is the second most preferred device for both genders.
10. Urban children have higher access to portable devices (Smartphones, Laptops).
11. Rural regions show greater reliance on TVs.
12. Urban kids show higher total screen exposure.
13. Physical health issues (eye strain, neck pain) are most common among Teenagers.
14. Mental health issues (stress, anxiety) rise in Late-teens.
15. Pre-teens show early health impact signs due to excess smartphone use.
16. Higher screen time directly correlates with health impact risk.
17. Teenagers exceed screen time limit most, followed by Pre-teens.
18. Late-teens are somewhat controlled, possibly due to academic or exam discipline.
19. Laptops have the highest educational usage ratio.
20. Smartphones are mainly used for entertainment, not study.
21. TV usage is mostly recreational or passive learning.
22. Tablets fall in between but lean toward gaming and videos.
```

```python
# Summary by Screen Time Category
screen_time_driver = df.groupby("Device_Category").agg(
    Avg_Age=("Age", "mean"),
    Avg_Screen_Time=("Avg_Daily_Screen_Time_hr", "mean"),
    Exceeded_Rate=("Exceeded_Recommended_Limit", "mean"),
    Avg_Ratio=("Educational_to_Recreational_Ratio", "mean"),
    Count=("Age", "count")
).round(2)

print("Screen_Time_Category Driver Summary:")
display(screen_time_driver)

# Summary by Health Category
health_driver = df.groupby("Health_Impact_Category").agg(
    Avg_Screen_Time=("Avg_Daily_Screen_Time_hr", "mean"),
    Exceeded_Rate=("Exceeded_Recommended_Limit", "mean"),
    Avg_Ratio=("Educational_to_Recreational_Ratio", "mean"),
    Count=("Age", "count")
).round(2)

print("Health_Category Driver Summary:")
display(health_driver)

# Summary by Age Band + Urban/Rural
age_location_driver = df.groupby(["Age_Band", "Urban_or_Rural"]).agg(
    Avg_Screen_Time=("Avg_Daily_Screen_Time_hr", "mean"),
    Avg_Ratio=("Educational_to_Recreational_Ratio", "mean"),
    Exceeded_Rate=("Exceeded_Recommended_Limit", "mean"),
    Count=("Age", "count")
).round(2)

print("Age Band + Location Driver Summary:")
display(age_location_driver)

# Summary by Gender + Primary Device
gender_device_driver = df.groupby(["Gender", "Primary_Device"]).agg(
    Avg_Screen_Time=("Avg_Daily_Screen_Time_hr", "mean"),
    Exceeded_Rate=("Exceeded_Recommended_Limit", "mean"),
    Avg_Ratio=("Educational_to_Recreational_Ratio", "mean"),
    Count=("Age", "count")
).round(2)

print("Gender + Primary Device Driver Summary:")
display(gender_device_driver)

# Summary by Educational to Recreational Ratio Category
df['Usage_Purpose_Category'] = pd.cut(df['Educational_to_Recreational_Ratio'],
                                      bins=[0,0.3,0.6,1],
                                      labels=["Recreation Heavy","Balanced Use","Education Heavy"])

ratio_driver = df.groupby("Usage_Purpose_Category").agg(
    Avg_Screen_Time=("Avg_Daily_Screen_Time_hr", "mean"),
    Exceeded_Rate=("Exceeded_Recommended_Limit", "mean"),
    Count=("Age", "count")
).round(2)

print("Usage Purpose Category Driver Summary:")
display(ratio_driver)

# Summary by Exceeded Limit Status + Device Category
limit_device_driver = df.groupby(["Exceeded_Recommended_Limit", "Device_Category"]).agg(
    Avg_Screen_Time=("Avg_Daily_Screen_Time_hr", "mean"),
    Avg_Ratio=("Educational_to_Recreational_Ratio", "mean"),
    Count=("Age", "count")
).round(2)

print("Exceeded Limit + Device Category Driver Summary:")
display(limit_device_driver)

#  Correlation Summary of Numeric
numeric_driver_corr = df[[
    "Age","Avg_Daily_Screen_Time_hr","Educational_to_Recreational_Ratio"
]].corr().round(2)

print("Numeric Driver Correlation Summary:")
display(numeric_driver_corr)
```

```python
# Insights : 
1. High screen time users show the highest rate of exceeding recommended limits.
2. Low screen time group has a better educational usage ratio and healthier habits.
3. Children reporting Physical + Mental health issues show maximum screen time.
4. Those with No health impacts maintain screen time mostly within healthy limits.
5. Urban Teenagers have the highest average screen time.
6. Rural Pre-teens spend the least time on screens and show lower negative health impact risk.
7. Boys using smartphones have maximum screen time.
8. Girls using laptops show more balanced screen usage for education.
9. Recreation-heavy users have the highest screen hours.
10. Education-heavy users spend moderate time and rarely exceed limits. 
11. Most overuse cases are linked to portable devices, especially smartphones.
12. Non-portable devices correlate with limited screen time discipline.
13. Screen time has a moderate positive correlation with age.
14. Higher screen time → lower educational usage ratio (negative learning impact).
```
