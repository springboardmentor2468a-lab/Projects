```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('/content/Indian_Kids_Screen_Time.csv')

seg_summary = df.groupby(['Age', 'Gender'])['Avg_Daily_Screen_Time_hr'].agg(['mean', 'count']).reset_index()
seg_summary = seg_summary.sort_values('mean', ascending=False)

# Visualization
plt.figure(figsize=(10,6))
sns.barplot(data=seg_summary, x='Age', y='mean', hue='Gender', palette='coolwarm')
plt.title("Average Screen Time by Age Group & Gender", fontsize=14)
plt.ylabel("Average Screen Time (hrs)")
plt.xlabel("Age")
plt.xticks(rotation=45)
plt.show()
```

```python
device_stats = df.groupby('Primary_Device')[['Avg_Daily_Screen_Time_hr','Educational_to_Recreational_Ratio']].mean().reset_index()

fig, ax1 = plt.subplots(figsize=(10,6))
sns.barplot(data=device_stats, x='Primary_Device', y='Avg_Daily_Screen_Time_hr', color='skyblue', ax=ax1)
ax1.set_ylabel("Avg Screen Time (hrs)")
ax1.set_xlabel("Primary Device")
ax1.set_title("Device Type vs Screen Time & Educational Balance")

# Add secondary axis for Edu/Recreation Ratio
ax2 = ax1.twinx()
sns.lineplot(data=device_stats, x='Primary_Device', y='Educational_to_Recreational_Ratio', color='red', marker='o', ax=ax2)
ax2.set_ylabel("Edu/Recreation Ratio")
plt.show()
```

```python
health_location = pd.crosstab(df['Urban_or_Rural'], df['Health_Impacts'], normalize='index') * 100

# Visualization
health_location.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.title("Health Impact Distribution by Urban/Rural Segment", fontsize=14)
plt.ylabel("Percentage of Users (%)")
plt.xlabel("Location Type")
plt.legend(title="Health Impact Category", bbox_to_anchor=(1.05,1))
plt.show()
```

```python
# Correlation matrix
corr = df[['Avg_Daily_Screen_Time_hr','Educational_to_Recreational_Ratio']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Screen Time and Edu/Recreation Ratio")
plt.show()
```

Segment-Based Insights

1. Age as the Key Divider
Screen time increases steadily with age. The biggest jump occurs when children move from pre-teen to teenage years, and usage stays high through late teens.

Pre-teens (<13 years): Spend about 4.16 hours daily. Their time is split across several devices, mainly TVs and tablets, with more focus on educational content compared to older groups.

Teenagers (13–17 years): Average 4.50 hours per day. Smartphones dominate usage, and the balance shifts toward entertainment and recreation.

Late-teens (18+ years): Record the highest average at 4.53 hours daily. Their screen time revolves around personal devices like smartphones and laptops.

2. Device Type Influences Behavior
The kind of device used shapes both how long and how content is consumed.

Smartphones: Used by nearly half the respondents and linked with high screen time across all age brackets, especially among teens.

TVs and Laptops: Teenagers using TVs and late-teens using laptops/tablets exceed 4.5 hours daily, indicating a preference for long-form media such as movies, shows, or gaming.

Tablets: Though less widely used, tablets correspond to the highest average screen time in late-teens (4.56 hours). For pre-teens, they’re more commonly used for educational purposes.

3. Subtle Demographic Differences
While gender and location don’t majorly influence screen time, minor variations exist.

Gender: Boys average 4.38 hours, slightly above girls at 4.33 hours.

Location: Rural children (4.37 hours) show marginally higher usage than urban peers (4.34 hours), possibly due to fewer offline activity options.

Factors Behind High Screen Time and Elevated Concern

1. Screen Time Itself as the Core Driver
The strongest link to higher concern levels is simply the amount of time spent on screens. As total daily hours rise, so does the concern level:

No Harm: 0.88 hrs/day

Attention Needed: 4.26 hrs/day

High Concern: 6.91 hrs/day

Extreme Concern: 11.18 hrs/day

This clear progression shows that screen duration directly determines the level of risk and concern.

2. Device Access and Portability
Age and device ownership together shape excessive usage. Teenagers and late-teens using personal portable devices—especially smartphones and laptops—are the most vulnerable. The ease of access and mobility of these devices make it easy for screen time to extend throughout the entire day.

3. Type of Content Consumed
While content type (educational vs. recreational) is a secondary factor, it still influences concern levels. Interestingly, those in the “Extreme Concern” group show a higher ratio of educational to recreational use (0.52). This suggests that even when screen time appears “productive,” excessive duration itself can be harmful, or that content labeled as educational may often overlap with recreational use.

4. Age-Based Social and Academic Influences
The sharp rise in screen use during the teenage years is driven by social and educational expectations. Increased participation in social media, online gaming, and digital learning tools reinforces long hours of device interaction, making extended screen time seem normal or even necessary.

Main Factors Influencing Screen Time Usage

1. Demographic and Environmental Influences

Age: The strongest predictor of screen use. Screen time increases steadily with age, peaking in the late-teen years. The data clearly shows this upward trend, consistent with global findings—children aged 8–12 average about 5 hours daily, while teenagers (13–18) spend roughly 7.5 hours.

Location: A clear urban–rural gap exists. Urban children spend much more time on screens due to greater device availability, internet access, and lifestyle differences compared to rural peers.

Gender: Boys generally record higher screen time than girls, and the difference grows with age. Among late-teens (17–18 years), males display both a higher median usage and a broader range of screen activity.

2. Device Preferences and Content Habits

Primary Device: Smartphones are the dominant device across all age brackets, followed by laptops. Teenagers using TVs and late-teens using portable devices (smartphones, laptops) exhibit the highest average screen hours, reflecting greater flexibility and more engagement opportunities.

Content Type: Entertainment-driven use (social media, gaming, streaming) tends to dominate, though some educational use is observed among younger users—especially on tablets.

3. Behavioral and Health Dimensions

Health Effects: Prolonged screen exposure correlates with multiple health concerns. Poor sleep emerges as the most reported issue, followed by eye strain. Broader studies also associate heavy screen use with anxiety, depression, and irritability, underscoring the behavioral and emotional risks linked to excessive digital engagement.

Final Cohort Insights

High-Risk Cohorts:
Teenagers and late-teens represent the most vulnerable groups, with prolonged screen use driven by TV and smartphones. Risk intensifies as total daily hours rise, regardless of whether the content is educational or recreational.

Top Cohorts by Usage:

Teenagers (TV users): ~4.59 hours/day (735 users) — long-form video content dominates.

Late-teens (Tablet/Laptop users): ~4.56–4.55 hours/day — portable, personal devices sustain continuous engagement.

Teenagers (Smartphone users): ~4.51 hours/day (1,760 users) — confirms strong mobile-first habits.

Cross-Cutting Drivers:

Age progression: Screen time climbs steadily from pre-teen to late-teen years as autonomy and device access expand.

Content mix: Duration, not content type, is the primary risk factor — even users with more “educational” content show high concern when screen hours exceed healthy limits.

Risk Linkage:
Concern levels scale almost linearly with average screen hours:

No Harm: ≈ 0.88 h/day

Attention Needed: ≈ 4.26 h/day

High Concern: ≈ 6.91 h/day

Extreme Concern: ≈ 11.18 h/day
→ Reinforces daily screen time as a direct control point for wellbeing.

Health Impacts:
Poor sleep and physical fatigue are the most reported issues, cutting across all demographics — positioning sleep hygiene as the most effective mitigation lever.

Recommended Actions:

Target teenagers and late-teens using smartphones/laptops for focused interventions.

Set daily caps, turn off autoplay, and enforce device-free routines during study and bedtime.

Convert “educational binge” sessions into scheduled, time-boxed activities.

Promote outdoor and social alternatives to reduce discretionary screen use.

Encourage family-based digital rules (no screens during meals or in bedrooms) to regulate exposure and improve consistency.
