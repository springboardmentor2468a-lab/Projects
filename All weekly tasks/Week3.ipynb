# Univariate & Bivariate Visual Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ------------------ Univariate Analysis ------------------
# 1. Distribution of Daily Screen Time Hours
plt.figure(figsize=(8,5))
sns.histplot(data['Avg_Daily_Screen_Time_hr'], bins=20, kde=True)
plt.title("Distribution of Daily Screentime Hours")
plt.xlabel("Daily Screen Time (hours)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Distribution of Age Bands
plt.figure(figsize=(8,5))
sns.countplot(
    data=data,
    x='Age',
    hue='Age',
    order=data['Age'].value_counts().index,
    palette="Set2",
    legend=False
)
plt.title("Distribution of Age Bands")
plt.xlabel("Age Band")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Pie Chart  Device Usage
device_counts = data['Primary_Device'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    device_counts,
    labels=device_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("Set3", len(device_counts))
)
plt.title("Device Usage Distribution")
plt.show()
```

```python
# 4. Distribution of Device Usage
plt.figure(figsize=(8,5))
sns.countplot(
    data=data,
    x='Primary_Device',
    hue='Primary_Device',
    order=data['Primary_Device'].value_counts().index,
    palette="Set3",
    legend=False
)
plt.title("Primary Device Usage")
plt.xlabel("Device Type")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()
```

```python
# ------------------ Bivariate Analysis ------------------

# 5. Compare Screentime by Gender
plt.figure(figsize=(8,5))
sns.boxplot(
    data=data,
    x='Gender',
    y='Avg_Daily_Screen_Time_hr',
    hue='Gender',
    palette="coolwarm",
    legend=False
)
plt.title("Screentime by Gender")
plt.xlabel("Gender")
plt.ylabel("Daily Hours")
plt.show()
```

```python
# 6. Compare Screentime by Age Band
plt.figure(figsize=(10,6))
sns.boxplot(
    data=data,
    x='Age',
    y='Avg_Daily_Screen_Time_hr',
    hue='Age',
    palette="viridis",
    legend=False
)
plt.title("Screentime by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Daily Hours")
plt.show()
```

```python
# 7. Compare Screentime by Location Type (Urban vs Rural)
plt.figure(figsize=(8,5))
sns.barplot(
    data=data,
    x='Urban_or_Rural',
    y='Avg_Daily_Screen_Time_hr',
    hue='Urban_or_Rural',
    estimator='mean',
    palette="Set1",
    legend=False
)
plt.title("Average Screentime by Location Type")
plt.xlabel("Location Type")
plt.ylabel("Average Daily Hours")
plt.show()
```

```python
# 8. Heatmap  Average Screentime by Age Band and Gender
# Create a pivot table
heatmap_data = data.pivot_table(
    index='Age',
    columns='Gender',
    values='Avg_Daily_Screen_Time_hr',
    aggfunc='mean',
    observed=False  # removes FutureWarning
)

plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Average Screentime by Age Band and Gender")
plt.xlabel("Gender")
plt.ylabel("Age Band")
plt.show()
```

```python
# 9. Line Plot  Average Screentime across Age Bands
age_hours = data.groupby("Age",observed=True)['Avg_Daily_Screen_Time_hr'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=age_hours, x="Age", y="Avg_Daily_Screen_Time_hr", marker="o")
plt.title("Average Screentime Trend across Age Bands")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Hours")
plt.show()
```
