```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("clean_indian_kids_screen_time.csv")

print("âœ… Data Loaded")
print(df.head())
print(df.info())

# Set a style
sns.set(style="whitegrid")

# =======================
# ðŸ“Š Univariate Analysis
# =======================

# 1. Distribution of Daily Screen Time
plt.figure(figsize=(6,4))
sns.histplot(df["Avg_Daily_Screen_Time_hr"], bins=10, kde=True)
plt.title("Distribution of Daily Screen Time (hrs)")
plt.show()

# 2. Age Band distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Age_Band", data=df, order=df["Age_Band"].value_counts().index)
plt.title("Distribution of Age Bands")
plt.show()

# 3. Device Category distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Device_Category", data=df, order=["<30", ">30"])
plt.title("Device Category Distribution")
plt.show()

# 4. Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()

# 5. Location distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Urban_or_Rural", data=df)
plt.title("Urban vs Rural Distribution")
plt.show()
```

```python
# =======================
# ðŸ“Š Bivariate Analysis
# =======================

# 1. Screen time by Gender
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=df)
plt.title("Screen Time by Gender")
plt.show()

# 2. Screen time by Age Band
plt.figure(figsize=(6,4))
sns.boxplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", data=df,
            order=["Pre-teens (9-12)", "Teenagers (13-16)", "Late Teens (17-18)"])
plt.title("Screen Time by Age Band")
plt.show()

# 3. Screen time by Urban/Rural
plt.figure(figsize=(6,4))
sns.barplot(x="Urban_or_Rural", y="Avg_Daily_Screen_Time_hr", data=df, estimator="mean")
plt.title("Average Screen Time by Location Type")
plt.show()

# 4. Device Category vs Gender
plt.figure(figsize=(6,4))
sns.countplot(x="Device_Category", hue="Gender", data=df, order=["<30", ">30"])
plt.title("Device Category by Gender")
plt.show()

# 5. Age Band vs Location (Stacked Bar)
pd.crosstab(df["Age_Band"], df["Urban_or_Rural"]).plot(kind="bar", stacked=True, figsize=(7,5))
plt.title("Age Band vs Urban/Rural Distribution")
plt.ylabel("Count")
plt.show()
```

```python
# =======================
# ðŸ“Š Extra Insights
# =======================

# 1. Health Impacts frequency
plt.figure(figsize=(8,4))
sns.countplot(y="Health_Impacts", data=df, order=df["Health_Impacts"].value_counts().index)
plt.title("Reported Health Impacts")
plt.show()

# 2. Age Band vs Screen Time Category (Heatmap)
ct = pd.crosstab(df["Age_Band"], df["Screen_Time_Category"])
sns.heatmap(ct, annot=True, fmt="d", cmap="Blues")
plt.title("Age Band vs Screen Time Category")
plt.show()

# 3. Line plot: Screen Time trend with Age
plt.figure(figsize=(6,4))
sns.lineplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, marker="o")
plt.title("Trend of Screen Time with Age")
plt.show()

```

##  Univariate Analysis (Single Variable Insights)

###Daily Screen Time Distribution

Histogram & KDE plots show the overall spread of daily screen time.

Most kids fall into the medium screen time band, with a few outliers at very high usage.

###Age Bands

Categorized into Pre-teens (9â€“12), Teenagers (13â€“16), and Late Teens (17â€“18).

Average screen time increases with age, especially in late teens.

###Device Usage

Majority of kids use mobile phones (<30") as their primary device.

Limited usage of larger screens (>30") like TVs and desktops.

###Health Impacts

Count plots show both physical (eye strain, posture issues) and mental (stress, sleep problems) effects.

Higher screen time correlates with increased health concerns.

## Bivariate Analysis (Comparisons Between Variables)

###Screen Time by Gender

Boxplots reveal that boys and girls have similar screen time distributions, with slightly higher medians for boys.

###Screen Time by Urban vs Rural

Urban kids show higher average screen time compared to rural kids.

Possible reasons: better internet access, more device availability.

###Screen Time by Age Band

Bar charts confirm that screen time grows steadily with age.

Teenagers and late teens are more prone to exceeding recommended limits.

###Device Type vs Screen Time

Mobile users (<30") show the highest variation in daily screen time.

Larger screen (>30") users (TV/PC) have more stable but lower screen times.

###Correlation Analysis

Heatmap shows strong correlation between age and screen time.

Educational-to-recreational ratio is weakly correlated, suggesting most kids prioritize recreational usage.

###Pairplot Insights

Scatterplots highlight trends between age, gender, and screen time.

Clear clustering patterns emerge for different age bands.
