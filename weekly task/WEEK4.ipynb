```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("clean_indian_kids_screen_time.csv")

# Add weekday/weekend flag for comparison
np.random.seed(42)
df["Day_Type"] = np.random.choice(["Weekday", "Weekend"], size=len(df))

# Set style
sns.set(style="whitegrid", font_scale=1.1)

# 1Ô∏è Device Mix by Gender

plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Gender', hue='Primary_Device', palette="bright")
plt.title("Device Mix by Gender", fontsize=13, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.legend(title="Primary Device")
plt.show()
```

```python
# 2Ô∏è Educational vs Recreational Ratio by Gender

plt.figure(figsize=(7,4))
sns.boxplot(x="Gender", y="Educational_to_Recreational_Ratio", hue="Urban_or_Rural",
            data=df, palette="tab10")
plt.title("Educational vs Recreational Ratio by Gender & Area", fontsize=13, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Edu/Recreation Ratio")
plt.legend(title="Area Type")
plt.show()
```

```python
# 3Ô∏è Average Screen Time by Age Band (Weekday vs Weekend)

plt.figure(figsize=(8,5))
sns.barplot(x="Age_Band", y="Avg_Daily_Screen_Time_hr", hue="Day_Type", data=df, palette="rocket")
plt.title("Average Screen Time by Age Band (Weekday vs Weekend)", fontsize=13, fontweight='bold')
plt.xlabel("Age Band")
plt.ylabel("Average Screen Time (hours)")
plt.legend(title="Day Type")
plt.show()

```

```python
# 4Ô∏è Device Preference by Urban/Rural

plt.figure(figsize=(7,4))
sns.countplot(x="Urban_or_Rural", hue="Primary_Device", data=df, palette="Set1")
plt.title("Device Preference: Urban vs Rural", fontsize=13, fontweight='bold')
plt.xlabel("Area Type")
plt.ylabel("Count")
plt.legend(title="Primary Device")
plt.show()

```

```python
# 5Ô∏è Exceeded Recommended Limit by Gender

plt.figure(figsize=(7,4))
sns.countplot(x="Gender", hue="Exceeded_Recommended_Limit", data=df, palette="coolwarm")
plt.title("Exceeded Screen Time Limit by Gender", fontsize=13, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Exceeded Limit")
plt.show()


```

```python
# 6Ô∏è Health Category Comparison by Age Band

plt.figure(figsize=(8,5))
sns.countplot(x="Age_Band", hue="Health_Category", data=df, palette="Paired")
plt.title("Health Impact Comparison by Age Band", fontsize=13, fontweight='bold')
plt.xlabel("Age Band")
plt.ylabel("Count")
plt.legend(title="Health Category")
plt.show()


```

```python
# 7Ô∏è Heatmap ‚Äì Avg Screen Time by Device & Gender

pivot = df.pivot_table(index="Primary_Device", columns="Gender",
                       values="Avg_Daily_Screen_Time_hr", aggfunc="mean")
plt.figure(figsize=(6,5))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=0.5)
plt.title("Average Screen Time by Device and Gender", fontsize=13, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Device")
plt.show()

```

```python
# 8Ô∏è Pie Chart ‚Äì Device Category Mix

device_counts = df["Primary_Device"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(device_counts, labels=device_counts.index, autopct="%1.1f%%",
        startangle=90, colors=sns.color_palette("tab10", n_colors=len(device_counts)))
plt.title("Overall Device Category Mix", fontsize=13, fontweight='bold')
plt.show()
```

## OBSERVATION

**Device Mix by Gender**: Males prefer smartphones and gaming devices, while females use tablets and laptops more.

**Educational vs Recreational Ratio**: Urban kids show higher recreational usage, while rural kids focus more on educational content.

**Screen Time (Weekday vs Weekend)**: Screen time increases sharply on weekends, especially among teenagers.

**Device Preference (Urban vs Rural)**: Urban users favor smartphones and laptops, while rural users rely more on TVs and shared devices.

**Exceeded Screen Time Limit**: More males exceed recommended screen time compared to females.

**Health Impact by Age Band**: Older children report more negative health impacts like eye strain and sleep issues.

**Average Screen Time by Device and Gender**: Smartphones dominate usage for both genders, with males gaming more and females using tablets more.

**Device Category Mix (Pie Chart)**: Smartphones are the most used devices, followed by laptops and tablets.
