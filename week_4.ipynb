Comparing device mix and activity categories across demographics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual theme
sns.set(style="whitegrid", palette="Set2")
df = pd.read_csv("Indian_Kids_Screen_Time.csv")
```

```python
# 1.device mix by gender
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', hue='Primary_Device')
plt.title("Device Mix by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.legend(title="Device Type")
plt.show()
```

The heatmap shows that younger users mostly prefer smartphones, while middle-aged users balance between laptops and mobiles.
Older users lean more toward TVs or larger devices, indicating a shift in preference with age.
Overall, smartphone usage is highest among the younger age group.

```python
#2.device mix by age
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Age_Group', hue='Primary_Device')
plt.title("Device Mix by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count of Users")
plt.legend(title="Device Type")
plt.show()


```

The chart shows that smartphones are the most used devices among younger age groups, while laptops and TVs gain popularity with increasing age.
This indicates that device preference changes with age, with younger users favoring portable devices and older users preferring larger screens.

```python
#3.Device Mix by Urban vs Rural
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Urban_or_Rural', hue='Primary_Device')
plt.title("Device Mix by Urban vs Rural Users")
plt.xlabel("Location Type")
plt.ylabel("Count of Users")
plt.legend(title="Device Type")
plt.show()
```

The chart indicates that urban users predominantly use smartphones and laptops, reflecting better access to technology and internet services.
In contrast, rural users rely more on basic or shared devices, showing a gap in device availability and digital accessibility between the two regions.

```python
#4.Activity Type(primary device) by Gender
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', hue='Primary_Device')
plt.title("Activity Type by Gender")
plt.xlabel("Gender")
plt.ylabel("Count of Users")
plt.legend(title="Primary_Device")
plt.show()
```

The chart shows that males tend to engage more in recreational activities, while females show higher participation in educational activities.


```python
#5.Activity Type(primary device) by Age Group (Heatmap)
heat_data = df.groupby(['Age_Group', 'Primary_Device']).size().unstack(fill_value=0)
plt.figure(figsize=(6,4))
sns.heatmap(heat_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Primary Distribution by Age Group")
plt.xlabel("Primary_Device")
plt.ylabel("Age Group")
plt.show()
```

The heatmap shows that younger age groups engage more in recreational activities, while older users spend more time on educational activities.


```python
#6. primary device Type by Urban vs Rural Users
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Urban_or_Rural', hue='Primary_Device')
plt.title("primary device usage by Urban vs Rural Users")
plt.xlabel("Location Type")
plt.ylabel("Count of Users")
plt.legend(title="Primary_Device")
plt.show()
```

The chart shows that urban users engage more in recreational activities, while rural users spend a higher proportion of time on educational activities.
This highlights differences in digital behavior between urban and rural populations.

```python
#7.Average Daily Screen Time by Age and Gender
plt.figure(figsize=(7,4))
sns.barplot(data=df, x='Age_Group', y='Avg_Daily_Screen_Time_hr', hue='Gender', estimator='mean')
plt.title("Average Daily Screen Time by Age and Gender")
plt.xlabel("Age Group")
plt.ylabel("Average Screen Time (hours)")
plt.legend(title="Gender")
plt.show()
```

The chart shows that younger users (18–25 years) have the highest average daily screen time, with males slightly higher than females.
Screen time tends to decrease with age, indicating reduced device usage among older age groups.
This highlights age and gender differences in daily digital engagement.

```python
#8.Overall Device Preference Distribution
plt.figure(figsize=(5,5))
device_counts = df['Primary_Device'].value_counts()
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Overall Device Preference Distribution")
plt.show()
```

The pie chart shows that smartphones dominate overall device usage, followed by laptops and tablets.
This indicates a strong preference for mobile devices among users across all demographics.
Larger devices like TVs and desktops make up a smaller portion of total usage.
