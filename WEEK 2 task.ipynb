```python
# Week 2: Preprocessing & Feature Engineering

import pandas as pd
import numpy as np

# Loading dataset
df = pd.read_csv("Indian_Kids_Screen_Time.csv")

# Displaying first 5 rows
df.head()
```

```python
# Replacing NaN values in Health_Impacts with "No Health Impact"
df["Health_Impacts"] = df["Health_Impacts"].fillna("No Health Impact")

df.head(10)
```

```python
import pandas as pd

# Function to classify age bands
def get_age_group(age):
    if pd.isna(age):
        return 'Missing'
    age = float(age)
    if 8 <= age <= 12:
        return 'Pre-teens'
    elif 13 <= age <= 16:
        return 'Teenagers'
    elif 17 <= age <= 18:
        return 'Late teens'
    else:
        return 'Outside range'

# Create Age_Group column
df['Age_Group'] = df['Age'].apply(get_age_group)

# Reorder columns to place Age_Group right after Age
cols = df.columns.tolist()
age_index = cols.index('Age')
# Remove Age_Group from the end
cols.remove('Age_Group')
# Insert Age_Group right after Age
cols.insert(age_index + 1, 'Age_Group')
df = df[cols]

# Show first 10 rows
print(df.head(10))

# Show counts per age group
print("\nCounts per Age Group:")
print(df['Age_Group'].value_counts())

df.head(5)
```

```python
import pandas as pd

# Defining keywords for classification
physical_keywords = ["eye strain", "obesity", "headache", "posture"]
mental_keywords = ["anxiety", "poor sleep", "stress", "depression", "attention"]

def categorize_health(impact):
    if pd.isna(impact) or impact.strip().lower() == "none":
        return "No health impact"
    
    text = impact.lower()
    has_physical = any(word in text for word in physical_keywords)
    has_mental = any(word in text for word in mental_keywords)
    
    if has_physical and has_mental:
        return "Physical & Mental"
    elif has_physical:
        return "Physical"
    elif has_mental:
        return "Mental"
    else:
        return "No Impact"

# Apply function
df["Health_Category"] = df["Health_Impacts"].apply(categorize_health)

# Reorder columns to place Health_Category right after Health_Impacts
cols = df.columns.tolist()
impact_index = cols.index("Health_Impacts")
cols.remove("Health_Category")
cols.insert(impact_index + 1, "Health_Category")
df = df[cols]

# Show first 10 rows
print(df[["Health_Impacts", "Health_Category"]].head(10))

# Show counts per Health Category
print("\nCounts per Health Category:")
print(df["Health_Category"].value_counts())

df.head(5)
```

```python
# Define device mappings
portable_devices = ["Smartphone", "Laptop", "Tablet"]
wallmounted_devices = ["TV"]

def get_device_category(device):
    if device in portable_devices:
        return "Portable"
    elif device in wallmounted_devices:
        return "Wallmounted"
    else:
        return "Unknown"

def get_screen_band(device):
    if device in portable_devices:
        return "<30"
    elif device in wallmounted_devices:
        return ">30"
    else:
        return "Unknown"

# Apply functions to generate new columns
df["Device_Category"] = df["Primary_Device"].apply(get_device_category)
df["Screen_Size_Band"] = df["Primary_Device"].apply(get_screen_band)

# Reorder columns: insert new columns after Primary_Device
cols = list(df.columns)
# Remove the two columns temporarily
cols.remove("Device_Category")
cols.remove("Screen_Size_Band")
# Find index of Primary_Device and insert after it
device_index = cols.index("Primary_Device")
cols.insert(device_index + 1, "Device_Category")
cols.insert(device_index + 2, "Screen_Size_Band")
# Reorder dataframe
df = df[cols]

# Show first few rows
print(df.head())

df.head(10)
```

```python
# Save updated dataset
df.to_csv("Indian_Kids_Screen_Time_Updated.csv", index=False)
```
